### Step 10:  Outlier Detection (IQR Method)

```python
q1 = df.quantile(0.25)

print(q1)



q3 = df.quantile(0.75)

print(q3)
```

```python
### To get the IQR of the data

iqr = q3 - q1 

print(iqr)
```

```python
### To find out the outlier

outliers = ((df < (q1 - 1.5 * iqr)) | (df > (q3 + 1.5 * iqr))).sum()
print(outliers)
```

```python
outliers_sorted = outliers.sort_values(ascending=False)

print(outliers_sorted)
```

### OBSERVATIONS:

1.  The main outliers lies in residual sugar, chlorides and the sulphates.
