### Bag Of Words

Bag Of Words is an imporatnt technique used in text representation in Natural Language Processing. It is used to consider every document as an input and convert it into the numerical vectors that ca be easily be trained by the Machine Learning Models.


### Importance of Bag Of Words :--

1.  Bag of Words is used to convert every text into the numerical vectors.

2.  The output of the Bag of words is in the form of 0 or 1.

3.  The output can easily be trained or understood by the Machine Learning Models.

4.  It is used for feature extraction

5.  It is used for smaller datasets


### Steps used in this Algorithm:---

1.  Import all the necessary libraries

2.  Import the object for CountVectorizer

3.  Define the Sample Text

4.  Transform the text into vector

5.  Convert vector into Numpy Array

6.  Get the columns from the vectors

7.  Get the DataFrame from the matrix

### Step 1: Import all the necessary libraries

```python
import   numpy   as np
import   pandas  as pd
import   matplotlib.pyplot  as  plt
import   seaborn            as  sns

from     sklearn.feature_extraction.text  import CountVectorizer
```

### Step 2: Import the object for CountVectorizer

```python
count = CountVectorizer()
```

### Step 3: Define the Sample Text

```python
corpus = [
    "I love data science and machine learning",
    "Data science is amazing",
    "I love coding in Python",
    "Machine learning and deep learning are related fields"
]
```

```python
corpus
```

### Step 4: Transform the text into vector

```python
X = count.fit_transform(corpus)
```

```python
X
```

### OBSERVATIONS:

1. The corpus text is converted into the Sparse Matrix

2. It has many non-zero values

3. It needs to be converted into the numpy array for better view and visibility

### Step 5: Convert vector into Numpy Array

```python
X = X.toarray()
```

```python
X
```

### OBSERVATIONS:

1. Now the sparse matrix has been converted into the numpy array for better view and visibility

### Step 6: Get the columns from the vectors

```python
print("Columns obtained from the vector is:", count.get_feature_names_out())
```

### OBSERVATIONS:

1. get_feature_names_out() -------> It learns all the unique words / vocabularies from the text.

### Step 7: Get the DataFrame from the matrix

```python
df = pd.DataFrame(X, columns = count.get_feature_names_out())
```

```python
df
```

### OBSERVATIONS:

1. The each and every corpus text is converted into the matrix representation format of zeros and ones.

2. Each sentence in the corpus is converted into a numerical vector representation.

3. Rows represent documents (sentences).

4. Columns represent unique words (vocabulary) extracted from the corpus.

5. Each cell value shows the frequency of a word in that document.

6. If a word is not present in a sentence → value is 0.

7. If a word appears once → value is 1.

8. If a word appears multiple times (e.g., "learning" in sentence 4) → value is greater than 1.

9. Word order and grammar are ignored — only word frequency matters.

10. The vocabulary is automatically converted to lowercase.

11. The output matrix is used as input features for Machine Learning models.
