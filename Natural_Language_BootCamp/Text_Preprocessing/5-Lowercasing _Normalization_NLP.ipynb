### Parts_Of_Speech_Tagging

Parts Of Speech (POS) refers to the porcess of assigning the parts of speech to each and every word in the sentence.

It is used to define the grammatical role for every word in the sentence.


### Imporatnce of Parts of Speech Tagging :- 

1. It is used to assign the labels to every word in the sentence and helps in defining the role for it.

2. It is used to identify the correct root word

3. It helps in understanding the type of the question and aswers the questions by understanding the sentence structure.

4. It helps in building the grammatical structure of the sentence by constructing the parse tree for it.





### OBSERVATION:


Ex:--  I am playing football.

                 I ----------------->   pronoun

                 am ---------------->   verb(auxillary)

                 playing ----------->   verb

                 football------------>  noun


Here parts of speech is assigned to every word in the particular text.

### Step 3:  Define the Input Text

```python
text = "The striped bats were hanging on their feet for best practices in Data Science!"
print("Original Text:\n", text)
```

### Step 4: Perform the Lowercasing on the entire text

```python
### Lower Case on the entire text

ans = text.lower()

print(ans)
```

### Step 5: Perform the Normalization (Remove Punctuation) on the entire text

```python
from nltk.tokenize  import RegexpTokenizer


### Create the object for Regular Expression Tokenizer

reg = RegexpTokenizer(r'\w+')

### using the object for Regular Expression, tokenize the text

answer = reg.tokenize(ans)

print(answer)
```

### Step 6:  Perform the Tokenization on the text

```python
### Convert the tokenized words in the list into strings

answer = " ".join(answer)

print(answer)

### perform the word tokeization on the text

words = word_tokenize(answer)

print(words)
```

### Step 7:  Perform the Lemmatization of the tokenized words

```python
from nltk.stem import WordNetLemmatizer


### Create the object for WordNetLemmatizer

lemma = WordNetLemmatizer()


res = [lemma.lemmatize(x) for x in words if(x != ' ')]

print(res)
```

### Step 8: Assign the Parts of Speech to every tokenized words

```python
### Assign the Parts of Speech to every word in the text

from   nltk  import pos_tag

postag = pos_tag(res)

print(postag)
```

### OBSERVATIONS:

1.  Here Parts of Speech has been assigned to every word in the text.

### Step 9: Perform the Lemmatization with POS

```python
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.stem  import WordNetLemmatizer

from  nltk.corpus import wordnet

### Create an empty list to store all the lemmatized words

lemmatized_words = []

for word,pos in postag:
    ### Convert the NLTK pos into WordNet pos
    respos = get_wordnet_pos(pos)
    lemmadata = lemma.lemmatize(word,pos=respos)
    lemmatized_words.append(lemmadata)

print(lemmatized_words)
```

### Step 10:  Plot the Lemmatized Words

```python
import matplotlib.pyplot as plt
from   collections import Counter

### Count the frequency of the lemmatized words

words_count = Counter(lemmatized_words)

print(words_count)


plt.figure(figsize=(8,6))
### perform the barplot of the frquency distribution of the words
plt.bar(words_count.keys(), words_count.values())

### set the title of the graph
plt.title("Frequency Distribution of the Lemmatized words")

plt.xlabel("Lemmatized Words")

plt.ylabel("Frequency Distribution")

plt.show()
```
