### Handling emojis & hashtags


#### Steps used in this Algorithm:

1.   Import all the Required Libraries

2.   Define the Sample Text (Social Media Data)

3.   Handling all the Emojis

4.   Create the Cleaned Text Pipeline

### Step 1:  Import all the Required Libraries

```python
import  re     ### used for regular expressions
import emoji   ### process the emojis
```

### Step 2:  Define the Sample Text (Social Media Data)

```python
text = "I love Data Science üòçüî• #MachineLearning #AI #DataScience2026 üöÄ"
print("Original Text:\n", text)
```

```python
print(text)
```

### OBSERVATIONS:

1.   Sample sentence contains:

       (a.)       Emojis ‚Üí üòçüî•üöÄ

       (b.)       Hashtags ‚Üí #MachineLearning, #AI, #DataScience2026

### Step 3: Handling all the Emojis

#### (i)  Convert Emojis to Text

```python
### Convert all the emojis to redeable text

text = emoji.demojize(text)

print(text)
```

#### (ii) Remove all the '#' (hastags)

```python
text = re.sub(r'#','',text)

print(text)
```

#### (iii) Remove all the extra spaces from the text

```python
text = re.sub(r'\s+',' ',text).strip()

print(text)
```

### Step 4: Create the Cleaned Text Pipeline

```python
### function definition
def clean_text(text):
    ### convert all the emojis into the readable text
    text = emoji.demojize(text)

    ### Remove all the '#' from the text
    text = re.sub(r'#','',text)

    ### Remobe all the empty spaces from the text
    text = re.sub(r'\s+',' ',text).strip()
   
    return(text)

### function call
ans = clean_text(text)

print("Cleaned Text is:", ans)
```
