## Chi-Square Test Statistics

Chi-Square Test Statistics is a statistical test used in Statistics. It is used in categorical data.

It is used to determine whether or not there is any significant difference between observed frequencies and expected frequencies in one or more categories of the statistical data.

## Types of Chi-Square Test

1.  Chi-Square Test of Independence

         It is a statistical used used for categorical data. It is a part of Chi Square test. It is used to determine whether thete is any strong association between the categorical variables or not

## Chi-Square Test of Independence

To determine whether there is any strong association between Gender and Shopping Preference

## Import all the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from scipy.stats import chi2_contingency
```

## Create sample data

```python
data = {'Shopping_Preference': ['Online', 'In-Store'],
        'Male': [60, 40],
        'Female': [30, 70]}

print(data)

df = pd.DataFrame(data)

print("------------------------------------Before Setting Index------------------------------------")
print(df)
```

```python
## Set the Shopping_Preference as index
df.set_index('Shopping_Preference', inplace=True)

print("------------------------------------After Setting Index------------------------------------")
print(df)
```

## Perform Chi-Square Test of Independence

```python
from scipy.stats import chi2_contingency

chi2,p,df,expected = chi2_contingency(df)


print("Chi Square Test Statistics :", chi2)

print("P value :", p)

print("Degree of Freedom :", df)

print("expected value :", expected)
```

## Interpret Results

```python
alpha=0.05

if(p<alpha):
    print("Alternate Hypothesis: There is no significant relationship between gender and shopping preference:")
else:
    print("Null Hypothesis: There is a significant relationship between the gender and shopping preference:")
```

## OBSERVATIONS:

(i) chi2_contingency is a statistical function imported from scipy.stats that is used to consider cross tabulated data
as input and performs the chi square test of independence on the two categorical data.

(ii) It checks whether there is any strong association between the two categories

(iii) It returns chi square test value, p value, degree of freedom amd the expected value.

2.  Chi-Square Goodness of Fit

         It is a statistical used used for categorical data. It is a part of Chi Square test. It is used to determine whether the observed data fits well in an expected distribution.

## Chi-Square Goodness of Fit

To check if the dice is fair.

## Import all the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import chisquare
```

## Observed vs Expected Frequencies

```python
## Observed frequencies (if the dice is rolled 10 times a day)

observed = np.array([8, 9, 10, 11, 12, 10])

## Expected frequencies(if the each face of the dice appears 10 times)

expected = np.array([10, 10, 10, 10, 10, 10])
```

## Perform Chi Square Goodness of Fit

```python
chi2_stat, p_val=chisquare(f_obs=observed, f_exp=expected)
```

```python
print("Chi- Square Statistics:", chi2_stat)
```

```python
print("P Value:", p_val)
```

```python
if(p_val < 0.05):
    print("Null Hypothesis : The dice is fair")
else:
    print("Alternate Hypothesis: The dice is biased")
```
