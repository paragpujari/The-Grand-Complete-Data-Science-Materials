## Statistics - Standard Normal Distribution

It is a kind of normal distribution where the mean is 0 and std is 1.


## Standard Normal Distribution

Normally distributed data can be transformed into a standard normal distribution.

Standardizing normally distributed data makes it easier to compare different sets of data.

The standard normal distribution is used for:

1.  Calculating confidence intervals

2.  Hypothesis tests


## Z-Values

It depicts how many standard deviation a value is away from the mean.

## Calculating the  zscore  value




```python
import numpy as np 

## define the dataset

x = np.array([10, 12, 14, 16, 18, 20])

print(x)


### Compute the mean and standard deviation from the above dataset

### mean of the above dataset

a = np.mean(x)

print("Mean is :", a)


### stadard deviation of the above dataset

b = np.std(x)

print("Standard Deviation is:", b)


### Calculating the Z-Score of the above dataset

score = (x - a)/b

print("Z-Score is:", score)
```

```python
## Fining out the Z Value or the zscore from the given dataset [10, 12, 14, 16, 18, 20]


import numpy as np

### define the dataset

x = np.array([10, 12, 14, 16, 18, 20])

print(x)

### Compute the zscore values from the above dataset


from scipy.stats import zscore


zdatavalues = zscore(x)

print("Z-Scores is:", zdatavalues)
```

```python
## With Python use the Scipy Stats library norm.ppf() function find the z-value separating the top 10% from the bottom 90%

from scipy import stats

### state the p_value

### As botton 90 % is the base, it is the cumulative probability, so p_value = 0.9

p_value = 0.9


z_value = stats.norm.ppf(p_value)

print(z_value)
```

```python
## With Python use the Scipy Stats library norm.cdf() function find the probability of getting less than a Z-value of 3


from scipy import stats

### state the z_value

z_value = 3

p_value = stats.norm.cdf(z_value)

print(p_value)
```
