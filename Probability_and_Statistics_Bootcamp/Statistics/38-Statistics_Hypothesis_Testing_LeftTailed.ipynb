## Statistics - Hypothesis Testing a Mean (Left Tailed)









```python
from scipy import stats
import math


## Step 1:-  Define the Hypothesis

### 1. Null Hypothesis(H0) :- Mean weight = 50 grms

### 2. Alternate Hypothesis(H1) :- Mean weight < 50 grms


## Step 2:-  Define the variables

x_hat = 50
n = 40
x_mean = 48
pop_std = 5

## Step 3:- Define the significance level

alpha = 0.05
alpha_half = alpha/2
alpha_half_new = 1 - alpha_half


## Step 4:- Compute the test statistics

res_num = (x_mean - x_hat)

res_den = pop_std/(math.sqrt(n))

res = res_num/res_den
print("Test Statistics is:", res)

## Step 5:- Compute the critical value

p_value = stats.norm.cdf(res)
print(p_value)

z_value = stats.norm.ppf(alpha)
print(z_value)

## Step 6:- Check the condition

if(res < z_value):
    print("Reject H0: Mean weight = 50 grms")
else:
    print("Accept H0: Mean weight = 50 grms")
```
