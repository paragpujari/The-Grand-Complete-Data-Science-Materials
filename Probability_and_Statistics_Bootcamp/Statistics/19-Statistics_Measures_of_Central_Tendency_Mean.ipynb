## Mean

The mean is the sum of all the values in the data divided by the total number of values in the data

The mean is calculated for numerical variables

A variable is something in the data that can vary, like:

Age

Height

Income

```python
## With Python use the NumPy library mean() method to find the mean of the values 4,11,7,14


import numpy as np

### dataset

data = [4,11,7,14]

### mean of the dataset

x = np.mean(data)

print(x)
```
