## Interquartile Range

Interquartile range is the difference between the first and third quartiles (Q1 and Q3)

```python
## With Python use the SciPy library iqr() method to find the interquartile range of the values 13, 21, 21, 40, 42, 48, 55, 72


from scipy import stats

### define the dataset

data = [13, 21, 21, 40, 42, 48, 55, 72]


###  specify the IQR of the dataset

iqrres = stats.iqr(data)

print(iqrres)
```

```python
## With Python use the SciPy library iqr() method to find the interquartile range of the values 13, 21, 42, 40, 56, 48, 87, 72, 98, 75


from scipy import stats

### define the dataset

data = [13, 21, 42, 40, 56, 48, 87, 72, 98, 75]

### depict the IQR from the dataset

res = stats.iqr(data)

print(res)
```
