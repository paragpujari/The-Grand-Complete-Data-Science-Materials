## Discrete probability distributions

1.   Bernoulli Distribution → Models single success/failure (Did user click ad: Yes/No)

2.   Binomial Distribution → Models number of successes in multiple trials (Clicks out of 10 ads)

3.   Poisson Distribution → Models number of events in fixed time (Customer arrivals per minute)

## 1. Bernoulli Distribution

```python
## import all the necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn  as sns
from   scipy.stats   import  bernoulli, binom, poisson
```

```python
## Generate the Bernoulli Distribution

p_click = 0.3
data_bernoulli = bernoulli.rvs(p_click, size=1000)
print("Bernoulli Distribution is:", data_bernoulli)
```

## OBSERVATION:

1.  From the bernoulli distribution, draw 1000 random samples whose probability of success is 30 %.

```python
## Generate the mean and variance from the random sample distribution

print("Mean of the sample distribution is:", np.mean(data_bernoulli))

print("Variance  of the sample distribution is:", np.var(data_bernoulli))
```

```python
## Display the bernoulli distribution graphically

sns.histplot(data_bernoulli, kde=True, color='red')
```

## 2. Binomial Distribution

```python
n_trials = 10   ## number of trials
p_success = 0.3 ## probability of success

## generate random samples from a binomial diribution

data_binom = binom.rvs(n_trials,p_success,size=1000)

print(data_binom)
```

## OBSERVATIONS:

1.  Here we have generated 1000 random data points from a binomial distribution for 10 different trials.

```python
print("Mean of binomial distribution", np.mean(data_binom))

print("Variance of binomial distribution", np.var(data_binom))
```

```python
## Display the binomial distribution graphically for 10 different trials

sns.histplot(data_binom, kde=True,color='orange',bins=10)
plt.title("Binomial Distribution: Number of Clicks out of 10 Ads")
plt.xlabel("Number of Clicks")
plt.ylabel("Frequency")
plt.show()
```

## 3. Poisson Distribution

```python
lambda_rate=4  ## average 4 customers arrive per minute

data_poisson = poisson.rvs(mu = lambda_rate, size=1000)
```

```python
data_poisson
```

## OBSERVATIONS:

1.  Here we are generating 1000 random data points from a poisson distribution with an average of 4 customers arriving per minute.

```python
print("Mean of Poisson Distribution is:",np.mean(data_poisson))

print("Variance of Poisson Distribution is:",np.var(data_poisson))
```

```python
## Display the Poisson Distribution graphically

sns.histplot(data_poisson,kde=True,color='orange',bins=10)
plt.title("Poisson Distribution: Customers Arrival per minute")
plt.xlabel("Number of Arrivals")
plt.ylabel("Frequency")
plt.show()
```
