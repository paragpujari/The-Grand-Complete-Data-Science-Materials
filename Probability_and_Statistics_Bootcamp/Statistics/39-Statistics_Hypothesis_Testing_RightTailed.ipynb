## Statistics - Hypothesis Testing a Mean (Right Tailed)

```python
from scipy import stats
import math

## Define the variables
pop_mean = 12
n = 36
sample_mean = 12.8
pop_std = 1.5
alpha = 0.05 

## Step 1: Define the Hypothesis Testing

### Null Hypothesis(H0) :- mean life battery = 12 hrs

### Alternate Hypothesis(H1) :- mean life battery > 12 hrs

## Step 2:- Define the Test statistics

res_num = sample_mean - pop_mean 

res_den = pop_std/math.sqrt(n)

res = res_num/res_den

print("Test Statistics is:", res)


## Step 3:- Define the critical value

p_value = stats.norm.cdf(res)
print("P Value is:", p_value)

z_value = stats.norm.ppf(1-alpha)
print("Z Critical value is:", z_value)

## Step 4:- Check the condition

if(res > p_value):
    print("Reject H0: mean life battery > 12 hrs")
else:
    print("Accept H0: mean life battery = 12 hrs")


print("----------or-----------------------")

if(res > z_value):
    print("Reject H0: mean life battery > 12 hrs")
else:
    print("Accept H0: mean life battery = 12 hrs")
```
