## Statistical Inference

It refers to the process of making conclusion from a given population.

There are two types of Statistical Inference:

(i)  Estimate

(ii) Hypothesis Testing


## Estimation

It refers to the process of using the sample data to predict the unknown population parameters.


The most likely value is Point Estimate


### Point Estimate

It refers to the single numerical value that is used to predict the unknown population parameter based on the sample data.


### Confidence intervals

It refers to the range of value that is used to estimate unknown population parameter with a certain level of confidence.


## Hypothesis Testing

It is used to check if the claim about the population is true.
