## Student's T Distribution

The Student's t-distribution (often just called the t-distribution) is a probability distribution used when estimating population parameters (like the mean) from a small sample.

It is especially when the population standard deviation (Ïƒ) is unknown.



## Finding out the p value from the t value:

```python
## With Python use the Scipy Stats library t.cdf() function find the probability of getting less than a t-value of 2.1 with 29 degrees of freedom

df=29
t=2.1

from scipy import stats

x = stats.t.cdf(t,df)

print(x)
```

## Finding the T-value of a P-Value

```python
## With Python use the Scipy Stats library t.ppf() function find the t-value separating the top 25% from the bottom 75% with 29 degrees of freedom


df=29
p=0.75

from scipy import stats

res = stats.t.ppf(p,df)

print(res)
```
