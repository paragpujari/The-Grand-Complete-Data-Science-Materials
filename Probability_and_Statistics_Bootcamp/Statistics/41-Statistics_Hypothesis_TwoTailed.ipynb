## Statistics - Hypothesis Testing a Proportion (Two Tailed)










```python
from scipy import stats 
import math

## Define all the variables
p0 = 0.6

p_hat = 0.52

n = 100 




## Step 1 :- Define the Hypothesis

### Null Hypothesis(H0) :- The sample proportion = population proportion

### Alternate Hypothesis(H1) :- The sample proportion != population proportion


## Step 2:- Define the conditions

res = n*p0
resnew = res>5



## Step 3:- Define the significance level

alpha = 0.05

alpha_half = alpha/2 

alpha_half_new = 1-alpha_half



## Step 4:- Compute the Test Statistics

res = (p_hat - p0)/math.sqrt((p0*(1 - p0))/n)



## Step 5:- Compute the critical value

z_critical = stats.norm.ppf(alpha_half_new)


z_critical_new = 2 * (1 - stats.norm.cdf(abs(res)))



print("Test Statistics is:", res)
print("Z Critical Value is:", z_critical)
print("Sample proportion is:", p_hat)

if(res < z_critical):
    print("Reject H0: Both the proportions are different:")
else:
    print("Accept H0: Both the proportion is same:")
```
