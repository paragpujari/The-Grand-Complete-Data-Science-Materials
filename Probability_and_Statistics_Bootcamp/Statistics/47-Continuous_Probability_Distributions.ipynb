## Understanding Continuous Distributions (Normal, Uniform, Exponential)

## 1. Normal Distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
```

```python
## perform a uniform distribution

normal_data = np.random.normal(loc=50, scale=10, size=1000)
print(normal_data)
```

```python
min(normal_data)
```

```python
max(normal_data)
```

```python
sns.histplot(normal_data,kde=True,color='blue')
```

## OBSERVATIONS:

1. A Gaussian distribution is displayed with a mean of 50 and standard deviation of 10.

2. A Gaussian distribution is obtained for the given series ranging from 17 till 88.

## 2. Uniform Distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

data_uniform = np.random.uniform(low=10, high=60, size=1000)
print(data_uniform)
```

```python
min(data_uniform)
```

```python
max(data_uniform)
```

```python
sns.histplot(data_uniform,kde=True,color='red')
```

## OBSERVATIONS:

1.  A uniform distribution is plotted with the numbers ranging from 10 till 60.

## 3. Exponential

```python
data_exponential = np.random.exponential(scale=10, size=10000000)
```

```python
data_exponential
```

```python
min(data_exponential)
```

```python
max(data_exponential)
```

```python
sns.histplot(data_exponential,kde=True,color='orange')
```

## OBSERVATIONS:

1.  Here an exponential distribution is plotted with the distribution number ranging between 3.37 till 163.
