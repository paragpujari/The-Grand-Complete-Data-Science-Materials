## Statistics - Hypothesis Testing a Mean (Two Tailed)

```python
from scipy import stats
import math

## Define the variables

pop_mean = 75
n = 50
sample_mean = 72
pop_std = 8



## Step 1:  Define the Hypothesis

### Null Hypothesis(H0):- mean score is same as 75

### Alternate Hypothesis(H1):- mean score is different than 75





## Step 2:- Define significance level

alpha = 0.05

alpha_half = alpha/2

alpha_half_new = 1 - alpha_half




## Step 3:- Calculate the Test Statistics

res_num = (sample_mean - pop_mean)

res_den = pop_std/(math.sqrt(n))

res = res_num/res_den

print("Tes Statistics is:", res)



## Step 4:- Calculate the critical value

z_critical_value = stats.norm.ppf(alpha_half_new)
print("Z Critical Value is:", z_critical_value)

## Check the condition

if((res < z_critical_value) or (res > z_critical_value)):
    print("Reject H0: mean score is same as 75")
else:
    print("Accept H0: mean score is same as 75")
```
