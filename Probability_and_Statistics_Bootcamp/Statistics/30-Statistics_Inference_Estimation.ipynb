## Statistics - Estimation

Point Estimate

It is a numerical value that helps in estimating an unknown population parameter based on the sample data.

The point estimate depends on the type of data:

1.   Categorical data: the number of occurrences divided by the sample size.

2.   Numerical data: the mean (the average) of the sample.



## Confidence Intervals

It is defined by a lower bound and a upper bound.

It refers to the range of values that is used to estimate an unknown population parameter based on the level of confidence.



## Confidence Levels

The confidence level is the probability that a confidence interval contains the true population parameter if we were to repeat the sampling process many times.

It is usually expressed as a percentage



## Margin of Error

It refers to the difference between the point estimate and upper and lower bounds.


Margin_of_error = Point_Estimate - Confidence_Interval




## Steps for Calculating the Confidence Interval

The following steps are used to calculate a confidence interval:

1.  Check the conditions

2.  Find the point estimate

3.  Decide the confidence level

4.  Calculate the margin of error

5.  Calculate the confidence interval
