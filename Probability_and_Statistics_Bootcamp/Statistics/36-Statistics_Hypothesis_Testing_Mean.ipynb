## Hypothesis Testing

It refers to the process of making claims about the population parameters based on the sample data.

Two Hypothesis:

(i)  Null Hypothesis ---->  The claim for which the testing is done.

(ii) Alternate Hypothesis ----> The opposite of the claim .



### Hypotheses for Testing a Mean

Suppose we want to test whether the mean weight of apples is 150g

Hypothesis:-

(i) Null Hypothesis(H0) :- The mean weight of the apples is 150 g.

(ii) Alternate Hypothesis(H1) :- (Two Tailed) The mean weight of the apples is not 150g.

(iii) Alternate Hypothesis(H1) :- (Left Tailed) The mean weight of the apples is less than 150g.

(iv) Alternate Hypothesis(H1) :- (Right Tailed) The mean weight of the apples is more than 150g.

## Test Statistic

The test statistic depends on whether the population standard deviation (σ) is known or unknown.

Case 1: Population σ is known → Z-test





Case 2: Population σ is unknown → T-test





## Steps for Hypothesis Testing a Mean

1.  Set hypotheses

     .   Decide H₀ -  Null Hypothesis
     
     .   Decide H1 -  Alternate Hypothesis 


2.  Choose significance level (α)

     .  Usually 0.05 (5%) or 0.01 (1%)


3.  Calculate test statistic

     .  Use Z or t formula

4.  Find critical value

    .  From Z-table or t-table, depending on α and tail type


5.  Decision rule

     .   Reject H₀ if |test statistic| > critical value (two-tailed)

     .   Reject H₀ if test statistic > critical value (right-tailed) or < -critical value (left-tailed)


6.  Conclusion

     .   State in simple words whether evidence supports H₁ or not

## Z-Test


Sample of 25 apples, sample mean weight = 152g, population std = 4g. Test at α = 0.05 if mean weight ≠ 150g.  Perform using Z-Test ?



Solution:-


Z-Test is performed on the above problem statement because the population standard deviation is given here.

```python
## Step 1:-  Define the Hypothesis

##  1.  Null Hypothesis :-  The population mean weight is equals to 150 grams.

##  2. Alternate Hypothesis :- The population mean weight is different than 150 grams.

## Step 2:-  Define the Variables

n = 25
sample_mean_x = 152
population_mean_x = 150
population_std = 4

## Step 3:-  Define the Significance Level

alpha = 0.05
alpha_half = alpha/2
# print(alpha_half)

from scipy import stats
import math 

## Step 4 :- Calculate the test statistics for the z test.

test_statistics_res = (sample_mean_x - population_mean_x)/(population_std/math.sqrt(n))

## Step 5 :- Calculate the critical value for two tailed test
z_critical = stats.norm.ppf(1-alpha_half)


print("Test Statistics is:", test_statistics_res)
print("Critical value for z test is:", z_critical)


if(test_statistics_res > z_critical):
    print("Reject H0:- The population mean weight is equals to 150 grams")
else:
    print("Failed to Reject H0:- The population mean weight is equals to 150 grams")
```

## T-Test


Sample of 25 apples, sample mean weight = 152g, sample std = 4g. Test at α = 0.05 if mean weight ≠ 150g.  Perform using T-Test ?



Solution:-


T-Test is performed on the above problem statement because the population standard deviation is not given here.

```python
from scipy import stats
import math

## Define the variables

n = 25
sample_mean_weight = 152
sample_std = 4
population_mean_weight = 150
df = n-1

## Step 1: -  Define the Hypothesis

### Null Hypothesis(H0):- The mean weight of all 25 apples is equals to  150 grams. 

### Alternate Hypothesis(H1):- The mean weight of all 25 apples is not equals to  150 grams. 





## Step 2:- Define the significance level

alpha_half = alpha/2 

alpha_half_new = 1 - alpha_half


### Step 3:- Define the T-Test Statistics

t_statistics_value = (sample_mean_weight - population_mean_weight)/(sample_std/math.sqrt(n))
print(t_statistics_value)


###  Step 4:- Define the crtitical t value
t_critical_value = stats.t.ppf(alpha_half_new,df)
print(t_critical_value)


if(t_statistics_value > t_critical_value):
    print("Reject H0: The mean weight of all 25 apples is equals to  150 grams")
else:
    print("Failed to reject H0: No sufficient data is provided.")
```
