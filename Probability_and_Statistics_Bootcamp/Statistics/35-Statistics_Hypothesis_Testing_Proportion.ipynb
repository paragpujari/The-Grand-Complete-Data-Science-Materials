## Statistics - Hypothesis Testing a Proportion

Hypothesis testing for a proportion checks whether the proportion in a population is equal to a specific value or different from it


```python
from scipy import stats
import math

## Step 1:  Define the Hypothesis

### 1.  Null Hypothesis(H0):-  The sample proportion is equals to population proportion

### 2. Alternate Hypothesis(H1):- The sample proportion differs from that of population proportion.

## Define the variables

population_proportion = 0.6
sample_proportion = 0.54
alpha = 0.05
alpha_half = alpha/2
n = 100

## Step 2: Define the conditions for size

res = n*population_proportion

res = (res>5)

print(res)

## Step 3: Calculate the Z-Test Statistics

ans = (sample_proportion - population_proportion)/math.sqrt((population_proportion*(1 - population_proportion))/n)

print(ans)

## Step 4: Calculate the Z- Critical value

z_critical_value = stats.norm.ppf(1-alpha_half)
print(z_critical_value)


if(ans > z_critical_value):
    print("Reject H0: The sample proportion is equals to population proportion")
else:
    print("Failed to reject H0: The sample proportion differs from that of population proportion")
```
