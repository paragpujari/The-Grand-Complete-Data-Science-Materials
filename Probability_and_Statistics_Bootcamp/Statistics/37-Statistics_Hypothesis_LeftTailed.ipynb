## Statistics - Hypothesis Testing a Proportion (Left Tailed)








```python
from scipy import stats
import math

## define the variables
n = 100
x = 40
p0 = 0.5
alpha = 0.05
p_hat = x/n

## Step 1 : Define the Hypothesis

###  1.  Null Hypothesis(H0): p_hat >= p0
###  2.  Alternate Hypothesis(H1) : p_hat < p0

## Step 2 : Define the significance level

alpha = 0.05
alpha_half = alpha/2
alpha_half_new = 1-alpha_half

## Step 3: Calculate the Test statistics

ans = (p_hat - p0)/math.sqrt((p0*(1-p0))/n)
print(ans)

## Step 4: Calculate the critical value

z_critical_value = stats.norm.ppf(alpha)
print(z_critical_value)

if(ans < z_critical_value):
    print("Reject the Null Hyothesis H0")
else:
    print("Failed to Reject Hypothesis H0")
```
