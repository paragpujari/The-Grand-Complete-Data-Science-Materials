## Normal Distribution

The normal distribution is described by the mean and the standard deviation.

The normal distribution is often referred to as a 'bell curve' because of it's shape:

1.  Most of the values are around the center 

2.  The median and mean are equal

3.  It has only one mode

4.  It is symmetric, meaning it decreases the same amount on the left and the right of the center


The area under the whole curve is equal to 1, or 100%


## Empirical Rule

1.  68 % of the data lie within 1 std of the average.

2.  95 % of the data lie within 2 std of the average.

3.  99 % of the data lie within 3 std of the average.


## Different Mean and Standard Deviations

Mean --> It describes where the center of the normal distribution is.

Standard Deviation --> It describes how spread out the normal distribution is
