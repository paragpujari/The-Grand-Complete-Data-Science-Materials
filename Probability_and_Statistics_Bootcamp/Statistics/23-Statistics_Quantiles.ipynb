## Quartiles

Quartiles are values that separate the data into four equal parts

```python
## With Python use the NumPy library quantile() method to find the quartiles of the values 13, 21, 21, 40, 42, 48, 55, 72

## Find out the quartiles from the given dataset


import numpy as np


x = [13, 21, 21, 40, 42, 48, 55, 72]

## print(x)


### find out the quartiles from the dataset

x = np.quantile(x,[0, 0.25, 0.5, 0.75, 1.0])

print(x)
```

## 1st Quartile

```python
### Find out the first Quartile from the dataset


first = np.quantile(x,0.25)
```

```python
print(first)
```

## 3rd Quartile

```python
### Find out the 3rd Quartile from the dataset

third = np.quantile(x,0.75)
```

```python
print(third)
```

## IQR

```python
### Find out the IQR from the dataset

ans = third - first
```

```python
print(ans)
```

```python
### Find out the IQR from the dataset

from scipy import stats

res_data = stats.iqr(x)
```

```python
print(res_data)
```

```python
## With Python use the NumPy library quantile() method to find the quartiles of the values 13, 21, 42, 40, 56, 48, 87, 72, 98, 75, 76,23,43,67,90,99


import numpy as np


### define the dataset

data = [13, 21, 42, 40, 56, 48, 87, 72, 98, 75, 76,23,43,67,90,99]

print(data)


### define the quantiles from the dataset

ans = np.quantile(data,[0, 0.25, 0.5, 0.75, 1.0])

print(ans)


### get the first quartile from the dataset

firstans = np.quantile(data,0.25)

print("First Quartile is:", firstans)


### get the third quartile from the dataset

thirdans = np.quantile(data,0.75)

print("Third Quartile is:", thirdans)


### Get the IQR from the dataset

res = thirdans - firstans

print("IQR of the dataset is:", res)


from scipy import stats

print("Range of the dataset is:", stats.iqr(data))
```
