## Basic Probability



1.   Given a list of outcomes, calculate the probability of getting an even number

```python
### Total outcomes

outcomes = [1,2,3,4,5,6]

print(outcomes)

### probability of getting an even number

events = [x for x in outcomes  if(x % 2)== 0]


print(events)


### Probability of getting an even number

probability = len(events) / len(outcomes)

print(probability)
```

## Empirical Probability


2. Simulate tossing a coin 10,000 times and estimate probability of heads.

```python
import random

### Number of Trials

trials = 10000

### number of heads

heads = sum(1 for i in range(trials) if random.choice(['H','T'])=='H')


### Probability of heads

probability = heads/trials

print(probability)
```

## Conditional Probability


3.  Find P(A|B) using Python.

```python
import pandas as pd

data = {
    'Passed': [1,1,0,1,0,1],
    'Studied': [1,0,0,1,1,1]
}

df = pd.DataFrame(data)



### perform the conditional probability

p_b = df[df['Studied'] == 1].shape[0]

print(p_b)
```

```python
p_ab = df[(df['Studied'] == 1) & (df['Passed'] == 1)].shape[0]
```

```python
print(p_ab)
```

```python
cond_prob = p_ab/p_b
```

```python
print("Conditional Probability is:", cond_prob)
```
