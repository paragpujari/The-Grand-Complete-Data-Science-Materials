### Common Probability Distributions:

1.  Bernoulli Distribution

```python
from scipy.stats import bernoulli


ans = bernoulli.rvs(p=0.6,size=10000)

print(ans)
```

### OBSERVATIONS:

1. It generates the series of random variate samples from the bernoulli distribution.

```python
### Find the mean of the Bernoulli Distribution

x = ans.mean()
```

```python
print(x)
```

2. Binomial Distribution

```python
from scipy.stats import binom
```

```python
x = binom.pmf(3,n=10,p=0.5)
```

```python
print(x)
```

3. Poisson Distribution

```python
from scipy.stats import poisson
```

```python
x = poisson.pmf(4, mu=3)
```

```python
print(x)
```

4. Standard Normal Dstribution

```python
from scipy.stats import norm

x = norm.cdf(1.96) - norm.cdf(-1.96)
```

```python
print(x)
```
