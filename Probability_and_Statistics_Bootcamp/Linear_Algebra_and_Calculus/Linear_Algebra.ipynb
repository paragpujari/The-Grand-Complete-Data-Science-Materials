### Linear Algebra

```python
import numpy as np
import pandas as pd
```

```python
v1 = np.array([2, 3])

v2 = np.array([4, 1])
```

```python
print(v1)
```

```python
print(v2)
```

```python
### Vector Addition


resadd = v1 + v2


print("Vector Addition is :", resadd)
```

```python
### Vector Substraction

ressub = v1 - v2

print("Vector Substraction is:", ressub)
```

```python
### Vector Multiplication

resmul = v1 * v2

print("Vector Multiplication is:", resmul)
```

```python
### Magniture of a vector

x1 = np.linalg.norm(v1)

print("Magnitude of the first vector is:", x1)
```

```python
x2 = np.linalg.norm(v2)

print("Magniture of the second vector is:", x2)
```

###  Dot Product

```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

dot_pro_res = np.dot(v1, v2)

print("Dot Product of two vectors is:", dot_pro_res)
```

### Matrix Multiplication

```python
A = np.array([[1, 2],
              [3, 4]])
```

```python
print(A)
```

```python
B = np.array([[2, 0],
              [1, 2]])
```

```python
print(B)
```

```python
res = np.dot(A,B)
```

```python
print(res)
```

### Transpose of Matrix

```python
X = np.array([[1, 2, 3],
              [4, 5, 6]])
```

```python
print(X)
```

```python
print("Transpose of the matrix is:", X.T)
```

### Determinant of Matrix

```python
A = np.array([[1, 2],
              [3, 4]])



print(A)
```

```python
determinant = np.linalg.det(A)
```

```python
print("Determinant of the Matrix is:", determinant)
```

### Inverse of Matrix

```python
A = np.array([[1, 2],
              [3, 4]])



print("Original Matrix is :", A)
```

```python
res = np.linalg.inv(A)
```

```python
print("Inverse Matrix is:", res)
```

### Solving Linear Equations

```python
A = np.array([[3, 1],
              [1, 2]])

b = np.array([9, 8])


x = np.linalg.solve(A,b)

print("Solution is:", x)
```
