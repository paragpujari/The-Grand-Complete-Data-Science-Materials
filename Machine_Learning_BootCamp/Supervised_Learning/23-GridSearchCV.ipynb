### Step 4:  Divide the dataset into independent and dependent variables

```python
### independent variable

X = df.drop(columns='target', axis=1)

### dependent variable
Y = df['target']
```

### Step 5: Divide the independent and dependent variables into training and testing data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42,stratify=Y)
```

### Step 6:  Create the pipeline

```python
from sklearn.pipeline import Pipeline

### create an object for Pipeline

pipe = Pipeline(steps=[
    ('scaler' , StandardScaler())  ,
    ('model'  , LogisticRegression(max_iter=1000))
])
```

### OBSERVATIONS:

1. Pipeline is used to make sure that the scaling happens inside the cross validation.

2. It is done to avoid the data leakage.

### Step 7: Define Hyperparameter Grid

```python
params_grid = {
    'model__C' : [0.01, 0.1, 1, 10, 100]      ,
    'model__penalty' : ['l2']                 ,
    'model__solver'  : ['lbfgs','liblinear']
}
```

### OBSERVATIONS:

1. model__C  --------------> It controls the regularization strength.

    (a.) The smaller the C ----------> Stronger is the regularization and simpler is the model.

    (b.) The larger the C  ----------> Weaker is the regulatization and complex is the model.

2. model__penalty -------------> Here It specifies the regularization type.

    (a.) Here penalty is l2 -----> l2 is Ridge Regression. It prevents overfitting.

3. model__solver ---------------> It specifies the type of optimization algorithm used to solve the problem.

### Step 8: Apply the GridSearchCV

```python
from sklearn.model_selection  import GridSearchCV


### create an object for GridSearchCV

gridobj = GridSearchCV(
    estimator=pipe               ,
    param_grid= params_grid      ,
    cv = 5                       ,
    scoring='accuracy'           ,
    n_jobs=-1                    ,
    verbose=1       
)
```

### Step 9: Train the Model

```python
gridobj.fit(X_train, Y_train)
```

### OBSERVATIONS:

1.  The GridSearchCV model is trained with the help of training data using 'l2' regularization so as to prevent overfitting.

### Step 10: Best Parameters & Best Model

```python
### get the best parameters

best_parameters = gridobj.best_params_

print("Best parameters for this model is:", best_parameters)
```

```python
### best model

best_estimator = gridobj.best_estimator_

print("Best estimator for this model is:", best_estimator)
```

### Step 11: Evaluate the performance of the model

```python
### predict the model

Y_pred = best_estimator.predict(X_test)
```

```python
print(Y_pred)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report

ac = accuracy_score(Y_test, Y_pred)

print("Accuracy score for this model is:", (ac * 100.0))
```

```python
cm = confusion_matrix(Y_test, Y_pred)

print(cm)
```

```python
cr = classification_report(Y_test, Y_pred)

print(cr)
```

### Step 12: Check whether the breast cancer data is benign or Malignant

```python
## Predict whether the tumor is benign or malignant

sample = pd.DataFrame(X.iloc[[0]])



### predict the sample

predictions = best_estimator.predict(sample)[0]

print(predictions)
```
