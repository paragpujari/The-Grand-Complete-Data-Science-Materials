### Step 4: Split the dataset into independent and dependent variables

```python
X = df.drop(columns='target',axis=1)

Y = df['target']
```

### Step 5: Split independent and dependent variables into training and testing data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.25,random_state=42,stratify=Y)
```

### Step 7: Perform Hyper Parameter Tuning to imrpove the accuracy of the model

```python
#### Create a Pipeline

pipe = Pipeline(
    [
        ('scaler' , StandardScaler()),
        ('pca'    , PCA())           ,
        ('clf'    , SVC())
    ]
)
```

```python
pipe
```

```python
### Define the Hyper parameter tuning

param_grid = {
    'pca__n_components':[0.90,0.95,0.99],
    'clf__C':[0.1,1,10],
    'clf__kernel':['linear','rbf'],
    'clf__gamma':['scale','auto']
}
```

```python
param_grid
```

### OBSERVATIONS:


1.  `pca__n_components` ---------->  Parameter of PCA step         

2.  `clf__C`            ---------->  Parameter of SVM classifier   

3.  `clf__kernel`       ---------->  Type of SVM decision boundary 

4.  `clf__gamma`        ---------->  Kernel influence (for RBF)    


```python
### Perform the Hyper parameter tuning of the model

grid = GridSearchCV(
    pipe                  ,
    param_grid            ,
    cv = 5                ,
    scoring = 'accuracy'  ,
    n_jobs  = -1 
)
```

```python
### Using the object of Hyper parameter model, train the data

grid.fit(X_train, Y_train)
```

### Step 8: Best Model Evaluation

```python
### get the best model

best_model = grid.best_estimator_

print("Best Model is:", best_model)

### using the best model, predict the test data

Y_pred = best_model.predict(X_test)

print(Y_pred)
```

```python
### get the best parameters

print("Best Parameters:", grid.best_params_)
```

```python
print("Accuracy:", accuracy_score(Y_test, Y_pred)*100.0)
```

```python
print("\nConfusion Matrix:\n", confusion_matrix(Y_test, Y_pred))
```

```python
print("\nClassification Report:\n", classification_report(Y_test, Y_pred))
```
