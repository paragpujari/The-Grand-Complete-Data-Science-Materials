### Write a program based on Simple Linear Regression to predict the marks obtained by the student based on the number of hours studied.

#### Simple Linear Regression

1. The main goal of this project is to implement the end to end Simple Linear Regression of predicting the marks obtained by the student based on the number of hours studied.

2.  We also need to get the accuracy score of the model of how much accurate the new marks obtained by the student is to the original marks based on the new test data of the number of hours studied.

1.    Import all the necessary libraries

6.    Divide the independent and dependent variables into training and testing data

7.    Train the model

8.    Find the intercept and slope of the model

9.    Predict the model

10.   Find the best line of fit using the Visualization graph

11.   Find the accuracy score of the model

#### Step 1:   Import all the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model    import LinearRegression
from sklearn.metrics         import r2_score, mean_absolute_error, mean_squared_error, root_mean_squared_error
```

### Step 2: Load the Dataset

```python
### load the dataset from the url

url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"

### read the dataframe from the mentioned csv

df = pd.read_csv(url)

print(df)
```

```python
### Read the first five rows of the dataset

print(df.head())
```

```python
### Read the last five rows of the dataset

print(df.tail())
```

```python
### Get the shape of the dataset

print("Shape of the dataset is:", df.shape)
```

```python
### Get the total length of the records

print("Total length of the records is:", len(df))
```

```python
### Display all the information about the columns used in the dataset

print(df.info())
```

```python
### get all the columns used in the dataset

print(df.columns)
```

```python
## Display the statistical summary of the dataset

print(df.describe())
```

```python
## To check for the null values in the dataset

df.isnull().sum()
```

### OBSERVATIONS:

1. There are no NULL Values in the dataset.

###  Step 4:  Visualize the data

```python
plt.scatter(df['Hours'], df['Scores'], color='green')
plt.title("Hours vs Scores")
plt.xlabel("Number of Hours Studied")
plt.ylabel("Marks Obtained")
plt.show()
```

### OBSERVATIONS:

1. From the above graph, we can see that with the increase in number of hours studied, marks obtained by the students has also increased.

### Step 5 :  Divide the dataset into independent and dependent variables

```python
X = df[['Hours']]
```

```python
Y = df[['Scores']]
```

```python
X
```

```python
Y
```

### Step 6 : Divide the independent and dependent variables into training and testing data

```python
### divide the dataset into 80 % training and 20 % testing

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
### Shape of the input data

print("Shape of the training data is:", X_train.shape)
print("Shape of the testing data is:",  X_test.shape)
```

```python
### Shape of the output data

print("Shape of the training data is:", Y_train.shape)
print("Shape of the testing data is:",  Y_test.shape)
```

### Step 7 :  Train the model

```python
from sklearn.linear_model import LinearRegression

#### create an object for Linear Regression

obj = LinearRegression()

### using the object for Linear Regression, train the model

obj.fit(X_train, Y_train)
```

### Step 8 : Find the intercept and slope of the model

```python
print("Slope of the graph is:", obj.coef_[0])
print("Intercept of the graph is:", obj.intercept_)
```

### Step 9 : Predict the model

```python
Y_pred = obj.predict(X_test)
```

```python
print(Y_pred)
```

```python
print("--------------------------------------Actual Result---------------------------------------------------")

print(Y_test)
```

```python
print("--------------------------------------Predicted Result---------------------------------------------------")

print(Y_pred)
```

### OBSERVATIONS:

1.  From the above graph, we can see that the predicted result slightly varies from the actual result.

### Step 10 :  Find the best line of fit using the Visualization graph

```python
#### Visualize the testing data


### plot the scatter plot for the testing data points

plt.scatter(X_test, Y_test, color = 'green')

### perform the best line of fit for the predicted data points
plt.plot(X_test, obj.predict(X_test), color = 'red')

### give the title of the graph
plt.title('Regression Line - Testing Data')

### give the name for X-Axis
plt.xlabel("Hours Studied")

### give the name for Y-axis
plt.ylabel("Percentage Score")

### plot the legend of the graph
plt.legend()


### display the graph
plt.show()
```

## OBSERVATIONS:

1.  The green line shows the actual scores obtained by the student and the red line shows the best line of fit (the predicted marks obtained by the student)

2. The gap between the actual points and the best line of fit is the Residual or the error.

### Step 11:  Find the accuracy score of the model

```python
## Model evaluation

from sklearn.metrics import  r2_score, mean_absolute_error, mean_squared_error, root_mean_squared_error

print("Mean Absolute Error of the model is:", mean_absolute_error(Y_test, Y_pred))
print("Mean Squared Error of the model is:", mean_squared_error(Y_test, Y_pred))
print("Root Mean Squared Error of the model is:", root_mean_squared_error(Y_test, Y_pred))
print("R Squared value of the model is:", r2_score(Y_test, Y_pred)*100.0)
```
