### Step 4: Divide the dataset into independent and dependent variables

```python
### Independent variables

X = df[['Level']].values
```

```python
Y = df[['Salary']].values
```

### Step 5: Divide the data into training and testing data if available

### Step 6 : Train the model using Linear Regression

```python
from sklearn.linear_model import LinearRegression


lin = LinearRegression()


### train the model using LinearRegression

lin.fit(X,Y)
```

### Step 7 : Visualize the Linear Regression Results

```python
### plot the actual data points

plt.scatter(X,Y, color = 'red', label = 'actual data')

### plot the best line of fit

plt.plot(X, lin.predict(X), color = 'green', label = 'best fit')

## title of the graph
plt.title("Linear Regression Result")

## X Axis
plt.xlabel("Position Label")

## Y Axis
plt.ylabel("Salary")

## display the graph
plt.show()
```

### Step 8: Fit a Polynomial Regression Model and then train using SLR

```python
### Fit the Polynomial Regression Model

from sklearn.preprocessing import PolynomialFeatures


### Add a PolynomialFeatures of degree of '4'

poly = PolynomialFeatures(degree=4)

### use the polynomial object, transform the input

X_poly = poly.fit_transform(X)

print(X_poly)
```

```python
### train the model

from sklearn.linear_model import LinearRegression


lr_poly = LinearRegression()

lr_poly.fit(X_poly, Y)
```

#### OBSERVATIONS:

1.  Now the input has been transformed into degree = 4.

### Step 9:  Visualize Polynomial Linear Regression Results

```python
## plot the actual data points

X_grid = np.arange(min(X), max(X), step=0.1)
X_grid = X_grid.reshape(len(X_grid),1)

plt.scatter(X, Y, color = 'red', label = 'actual data')

## perform the best line of fit
plt.plot(X_grid, lr_poly.predict(poly.transform(X_grid)), color = 'green', label = 'Label Fit')

## title of the graph
plt.title("Polynomial Linear Regression Result")

## X Axis
plt.xlabel("Position Label")

## Y Axis
plt.ylabel("Salary")

## display the graph
plt.show()
```

### OBSERVATIONS:

1. Now a Polynomial Linear Regression curve has been created where a best fit of line exactly passes through the curved line of non-linear data.

### Step 10: Model Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error, r2_score

print("------------------------------Linear Data-----------------------------------------------")
Y_pred = lin.predict(X)

print(Y_pred)

print("------------------------------Polynomial Data-------------------------------------------")

Y_pred_poly = lr_poly.predict(X_poly)

print(Y_pred_poly)
```

```python
print("------------------------------Linear Data-----------------------------------------------")

mse = mean_squared_error(Y, Y_pred)
print("Mean Squared Error is:", mse)


mae = mean_absolute_error(Y, Y_pred)
print("Mean Absolute Error is:", mae)


rmse = root_mean_squared_error(Y, Y_pred)
print("Root Mean Squared Error is:", rmse)


r2score = r2_score(Y, Y_pred)
print("R2 Score is:", r2score)
```

```python
print("------------------------------Polynomial Data-----------------------------------------------")

mse = mean_squared_error(Y, Y_pred_poly)
print("Mean Squared Error is:", mse)


mae = mean_absolute_error(Y, Y_pred_poly)
print("Mean Absolute Error is:", mae)


rmse = root_mean_squared_error(Y, Y_pred_poly)
print("Root Mean Squared Error is:", rmse)


r2score = r2_score(Y, Y_pred_poly)
print("R2 Score is:", r2score)
```

### OBSERVATIONS:

1. From the above , we came to the conclusion that the accuracy of the model has increased a lot by adding the polynomial features or adding the degrees to the original linear data.
