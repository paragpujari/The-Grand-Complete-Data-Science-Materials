### Step 4: Divide the dataset into independent and dependent variables

```python
X = df.drop(columns='target', axis=1)

Y = df['target']
```

```python
print(X)
```

```python
print(Y)
```

### Step 5: Divide the independent and dependent variables into training and testing data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
print("Shape of the input training data is:", X_train.shape)

print("Shape of the input testing  data is:", X_test.shape)
```

```python
Y_test
```

```python
print("Shape of the output training data is:", Y_train.shape)

print("Shape of the output testing  data is:", Y_test.shape)
```

### Step 6:  Perform Feature Scaling on the input data

```python
from sklearn.preprocessing import StandardScaler

### create an object for Standard Scaler

sc = StandardScaler()

### using the object for Standard Scaler, scale the inputs

X_train_scaled = sc.fit_transform(X_train)

X_test_scaled  = sc.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

### OBSERVATIONS:

1.  All the inputs has been scaled in one range that makes it more easier for the training.

### Step 7:  Train the model

```python
from sklearn.svm import SVC

svcmodel = SVC(kernel = 'rbf', C = 0.1, gamma = 'scale')

### using the training data, train the model

svcmodel.fit(X_train_scaled, Y_train)
```

### OBSERVATIONS:

1. Now the svc model is called with the help of thefollowing parameters:- 

     (a.)   kernel = 'RBF' ---------------> It converts the data into the high dimensional space.

     (b.)    C = 0.1       ---------------> Regularization parameter that reduces the overfitting.

     (c.)    gamma = 'scale' -------------> It controls the influence of the single training point.


2. The model has been trained with the help of the training data using the following parameters:-

     (a.)  C ----------->  Regularization parameter

     (b.)  kernel = 'rbf' --------> Non linear data

     (c.)  gamma   --------------> controls the influence of the single training point.

### Step 8: Predict the model on the test data

```python
Y_pred = svcmodel.predict(X_test_scaled)
```

```python
Y_pred
```

### Step 9:  Evaluate the model performance

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("=========================Accuracy of the model============================")
ac = accuracy_score(Y_test, Y_pred)

print("Accuracy of the model is:", (ac * 100.0))

print("=========================Confusion Matrix of the model============================")
cm = confusion_matrix(Y_test, Y_pred)

print(cm)


print("=========================Classification Report of the model============================")
cr = classification_report(Y_test, Y_pred)

print(cr)
```

### Step 10: Visualize the confusion matrix

```python
import seaborn as sns

plt.figure(figsize=(10,5))

sns.heatmap(cm, annot=True)

plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - SVM Classifier')
plt.show()
```

### Step 11:  Predict whether the tumor is benign or malignant

```python
sample = X_test_scaled[0].reshape(1,-1)

res = svcmodel.predict(sample)

print(res[0])
```

```python
if(res[0] == 1):
    print("Brain Tumor is benign in nature:")
else:
    print("Brain Tumor is malignant in nature.")
```
