### Step 1: Import all the necessary libraries

### Step 6 : Train the model using Linear Regression

```python
from sklearn.linear_model import LinearRegression

lin = LinearRegression()

### fit and train the model

lin.fit(X,Y)
```

### Step 7: Visualize Linear Regression Results

```python
### plot the actual data points

plt.scatter(X,Y,color = 'red', label = 'actual data point')

### plot the best line of fit
plt.plot(X, lin.predict(X), color = 'green', label = 'best fit')

## title of the graph
plt.title("Linear Regression Result")

## X Axis
plt.xlabel("House Size")

## Y Axis
plt.ylabel("Price")

## display the graph
plt.show()
```

### Step 8: Fit a Polynomial Regression Model and then train using SLR

```python
### create a function for Polynomial Linear Regression

poly = PolynomialFeatures(degree=4)

### add the degree to the input
X_poly = poly.fit_transform(X)

print(X_poly)

print(X)
```

```python
### Train the model using the polynomial degree of the input

lin_poly = LinearRegression()

lin_poly.fit(X_poly,Y)
```

### Step 9: Visualize Polynomial Linear Regression Results

```python
## plot the actual data points

X_grid = np.arange(min(X), max(X), step=0.1)
X_grid = X_grid.reshape(len(X_grid),1)

plt.scatter(X, Y, color = 'red', label = 'actual data')

### perform the best line of fit
plt.plot(X_grid, lin_poly.predict(poly.transform(X_grid)),color = 'green', label = 'Label Fit')

## title of the graph
plt.title("Polynomial Linear Regression Result")

## X Axis
plt.xlabel("House Size")

## Y Axis
plt.ylabel("Price")

## display the graph
plt.show()
```

### OBSERVATIONS:

1.  From the above graph, we can see that with the increase in the size of the house, house price increases.

### Step 10:  Model Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error, r2_score

print("------------------------------Linear Data-----------------------------------------------")
Y_pred = lin.predict(X)

print(Y_pred)

print("------------------------------Polynomial Data-------------------------------------------")

Y_pred_poly = lin_poly.predict(X_poly)

print(Y_pred_poly)



print("------------------------------Linear Data-----------------------------------------------")

mse = mean_squared_error(Y, Y_pred)
print("Mean Squared Error is:", mse)


mae = mean_absolute_error(Y, Y_pred)
print("Mean Absolute Error is:", mae)


rmse = root_mean_squared_error(Y, Y_pred)
print("Root Mean Squared Error is:", rmse)


r2score = r2_score(Y, Y_pred)
print("R2 Score is:", r2score)



print("------------------------------Polynomial Data-----------------------------------------------")

mse = mean_squared_error(Y, Y_pred_poly)
print("Mean Squared Error is:", mse)


mae = mean_absolute_error(Y, Y_pred_poly)
print("Mean Absolute Error is:", mae)


rmse = root_mean_squared_error(Y, Y_pred_poly)
print("Root Mean Squared Error is:", rmse)


r2score = r2_score(Y, Y_pred_poly)
print("R2 Score is:", r2score)
```
