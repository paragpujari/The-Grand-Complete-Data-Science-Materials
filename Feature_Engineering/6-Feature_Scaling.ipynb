### 1. Feature Scaling

It is the process of making all the values in one particular range.

```python
import numpy as np

X = np.array([10, 20, 30, 40, 50])

print("Original Array is:", X)


### Find the mean value of X

mean_x = X.mean()

### Standard Deviation value of x

std_x = X.std()

print(X)

print(mean_x)

print(std_x)


scaled_X = (X - mean_x)/std_x

print(scaled_X)
```

### 2. Z-Score Scaling


It is the process of scaling all the values in one range using the mean and standard deviation.

```python
import pandas as pd

data = {
    'Age': [20, 25, 30, 35],
    'Salary': [20000, 50000, 80000, 120000]
}

df = pd.DataFrame(data)
print(df)
```

```python
from sklearn.preprocessing import StandardScaler

### Create an object for Standard Scaler

sc = StandardScaler()

### using the standard scaler transform the data

transformed_scaled_data = sc.fit_transform(df)

### print(transformed_scaled_data)


### Construct the dataframe using the scaled data

scaled_df = pd.DataFrame(transformed_scaled_data, columns = df.columns)

print(scaled_df)
```

### 3. Min-Max Scaling

Brings all values into same range [0,1]

```python
from sklearn.preprocessing import MinMaxScaler


### create an object for MinMaxScaler

minmaxscaler = MinMaxScaler()


### using the object of MinMaxScaler, transform the data

scaled_data = minmaxscaler.fit_transform(df)


### using the scaled data, construct the dataframe

scaled_df_minmax = pd.DataFrame(scaled_data, columns = df.columns)

print(scaled_df_minmax)
```

### 4. Robust Scaling

It is used when outliers exists.

It is not affected by the extreme values.

```python
from sklearn.preprocessing import RobustScaler


### create an object for RobustScaler

robscalerobj = RobustScaler()


### using the object, transform the data into the scaled data

robust_scaled_data = robscalerobj.fit_transform(df)

## print(robust_scaled_data)

### using the scaled data contract the dataframe

robust_scaled_df = pd.DataFrame(robust_scaled_data, columns = df.columns)

print(robust_scaled_df)
```

### 5.  MaxAbs Scaling

It gives the value in the range of -1 to 1.

```python
from sklearn.preprocessing import MaxAbsScaler


### create an object for RobustScaler

maxscaler = MaxAbsScaler()


### using the object, transform the data into the scaled data

max_abs_scaled_data = maxscaler.fit_transform(df)

## print(robust_scaled_data)

### using the scaled data contract the dataframe

max_abs_scaled_df = pd.DataFrame(max_abs_scaled_data, columns = df.columns)

print(max_abs_scaled_df)
```
