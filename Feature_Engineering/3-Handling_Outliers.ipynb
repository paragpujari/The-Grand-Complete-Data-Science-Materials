## 5 number Summary And Box Plot

```python
lst_marks=[45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74,500]
```

```python
### perform the boxplot for the above dataset

import seaborn as sns
```

```python
sns.boxplot(lst_marks)
```

### OBSERVATIONS:

1.  From the above graph, we can see that there is a boxplot and it has one outlier at point 500 very far away from the boxplot range.

### Perform 5 Number Summary

```python
### Perform the box plot

import numpy as np

minimum,Q1,median,Q3,maximum = np.quantile(lst_marks,[0, 0.25, 0.5, 0.75, 1.0])
```

```python
print("Minimum Value is:", minimum)

print("First Quantile Value is:", Q1)

print("Median Value is:", median)

print("Third Quantile Value is:", Q3)

print("Maximum Value is:", maximum)
```

```python
### To remove the outlier we need to define the lower and upper bound

##### lb = q1 - (1.5 * iqr)

##### ub = q3 + (1.5 * iqr)
```

```python
IQR = Q3 - Q1
```

```python
print("Inter Quartile Range:", IQR)
```

```python
lower_bound = Q1 - (1.5 * IQR)
```

```python
upper_bound = Q3 + (1.5 * IQR)
```

```python
print(lower_bound, upper_bound)
```

### OBSERVATIONS:

1. The range of the entire dataset lies between 1.5 and 141.5. 

2. As 500 lies outside the range, so it is considered an outlier. Thus this value is removed.
