## Constructors

It is a special method used in classes and objects that is automatically created once the object is created.

It is used to initailize all the member variables of the class.


## Types of Constructors

## 1.  Default Constructors

It is a constructor that do not have any parameter.


No parameter is passed to the member function while object accessing the member function inside the class.


Here only default values are passed to the member functions.


## __init__()

It is a constructor created inside the class that gets initialized once the object of the class gets created.

It is used to initialize the member variables of the class.

```python
## Write a program to calculate the area and perimeter of the rectangle


class Rectangle:
    ### initialize the length and breadth of the rectangle
    def __init__(self):
        self.length = 120
        self.breadth = 80

    ### print the details of the rectangle
    def print_details(self):
        print("Length of the rectangle is:", self.length)
        print("Breadth of the rectangle is:", self.breadth)

    
    ### print the area and perimeter of the reactangle
    def print_area_rec(self):
        return(self.length * self.breadth)

    def print_perimeter_rec(self):
        return(2*(self.length + self.breadth))


rec = Rectangle()
print("Area of the rectangle is:", rec.print_area_rec())
print("Perimeter of the rectangle is:", rec.print_perimeter_rec())
```

## 2. Parameterized Constructor

It is a type of constructor that has parameters

Here parameters are present in the member function inside the class.

Dynamic values are passed to the member function inside the class.

```python
## Write a program to calculate the area and perimeter of the rectangle

class Rectangle:
    def __init__(self,l,b):
        self.length = l
        self.breadth = b

    ### print the details of the rectangle
    def print_details(self):
        print("Length of the rectangle is:", self.length)
        print("Breadth of the rectangle is:", self.breadth)

    ### print the area of the rectangle
    def calculate_area_rec(self):
        return(self.length * self.breadth)

    def calculate_peri_rec(self):
        return(2*(self.length + self.breadth))

rec = Rectangle(120,120)
rec.print_details()
print("Area of the rectangle is:",      rec.calculate_area_rec())
print("Perimeter of the rectangle is:", rec.calculate_peri_rec())
```
