## Create Labels

With the index argument, you can name your own labels

```python
import pandas as pd 

a = [10,20,30]

# creating the Series with the help of labels

res = pd.Series(a, index=['a','b','c'])

print(res)
```

```python
## Return all the series values through label


print(res['a'])
print(res['b'])
print(res['c'])
```

```python
## Create a simple Pandas Series from a dictionary

import pandas as pd 

calories = {"day1": 420, "day2": 380, "day3": 390}

print(calories)

### series

ser = pd.Series(calories)

print(ser)


### Accessing the elements from the series

print(ser['day1'])
print(ser['day2'])
print(ser['day3'])
```

## Replace Using Mean, Median, or Mode

A common way to replace empty cells, is to calculate the mean, median or mode value of the column.

Pandas uses the mean() median() and mode() methods to calculate the respective values for a specified column

### 1. mean()

```python
import pandas as pd

df = pd.read_csv('data.csv')

print("-------------------Before Cleaning------------------------------------")

print(df.isnull().sum())

x = df['Calories'].mean()
print('Mean is :', x)

print("-------------------After Cleaning------------------------------------")

df['Calories'] = df['Calories'].fillna(df['Calories'].mean())

print(df.isnull().sum())
```

```python
df[df['Calories'] == 375.79024390243904]
```

## OBSERVATIONS:

1.  After filling all the NULL Values of Calories with the mean value of it, there are now no NULL Values for Calories column.

## 2. median()

```python
import pandas as pd

df = pd.read_csv('data.csv')

print("-------------------Before Cleaning------------------------------------")

print(df.isnull().sum())

x = df['Calories'].median()
print(x)


print("-------------------After Cleaning------------------------------------")

df['Calories'] = df['Calories'].fillna(x)
print(df.isnull().sum())
```

```python
df[df['Calories'] == 318.6]
```

## 3. mode()

```python
import pandas as pd

df = pd.read_csv('data.csv')

print("-------------------Before Cleaning------------------------------------")

print(df.isnull().sum())

x = df['Calories'].mode()[0]

print("Mode is:", x)

print("-------------------After Cleaning------------------------------------")

df['Calories'] = df['Calories'].fillna(x)

print(df.isnull().sum())
```

```python
df[df['Calories']==300.0]
```
