## JSON 

It stands for Java Script Object Notation.

It is used for exchanging and storing the data.

It has an inbuilt package called 'json' that is used to work with json data.

```python
import json

## json to python object

x = '{ "name":"John", "age":30, "city":"New York"}'

print(x)

## json to python object

ans = json.loads(x)

print(ans)
```

```python
print(type(ans))
```

```python
## json object

x = '{ "name":"John", "age":30, "city":"New York", "country":"USA"}'

print(x)
```

```python
## python object

ans = json.loads(x)

print(ans)
```

```python
print(type(ans))
```

```python
## Convert from Python Object to JSON String

import json

# python object

x = {
    'name' : 'Parag'  ,
    'age'  :  30      ,
    'city' : 'Sambalpur'
}


print(x)

print(type(x))
```

```python
## json object

res = json.dumps(x)
```

```python
print(res)
```

```python
print(type(res))
```

```python
## Convert Python objects into JSON strings, and print the values

print(json.dumps({"name": "John", "age": 30}))

print(type(json.dumps({"name": "John", "age": 30})))
```

```python
print(json.dumps([10,20]))

print(type(json.dumps([10,20])))
```

```python
print(json.dumps((10,20)))

print(type(json.dumps((10,20))))
```

```python
print(json.dumps("hello"))

print(type(json.dumps("hello")))
```

```python
print(json.dumps(42))

print(type(json.dumps(42)))
```

```python
print(json.dumps(31.76))

print(type(json.dumps(31.76)))
```

```python
print(json.dumps(True))

print(type(json.dumps(True)))
```

```python
print(json.dumps(False))

print(type(json.dumps(False)))
```

```python
print(json.dumps(None))


print(type(json.dumps(None)))
```

```python
## Convert a Python object containing all the legal data types

import json

x = {
    "name": "John",
    "age": 30,
    "married": True,
    "divorced": False,
    "children": ("Ann","Billy"),
    "pets": None,
    "cars": [
      {"model": "BMW 230", "mpg": 27.5},
      {"model": "Ford Edge", "mpg": 24.1}
    ]
}

print(x)
```

```python
## json object

ans = json.dumps(x)

print(ans)
```

```python
print(type(ans))
```

```python
resdata = json.dumps(x, indent=4)

print(resdata)
```

```python
resdata = json.dumps(x, indent=4,sort_keys=True)

print(resdata)
```

```python
resdata = json.dumps(x, indent=4, separators=(".","="))

print(resdata)
```
