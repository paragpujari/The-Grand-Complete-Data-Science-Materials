###  Python Conditions and If statements

Python supports the usual logical conditions from mathematics:

.     Equals: a == b

.     Not Equals: a != b

.     Less than: a < b

.     Less than or equal to: a <= b

.     Greater than: a > b

.     Greater than or equal to: a >= b

These conditions can be used in several ways, most commonly in "if statements" and loops

## 1. If Statements

```python
age = 18

if(age >= 18):
    print("You are eligible to vote:")
```

```python
numb = int(input('Enter the number:'))
if(numb > 0):
    print('Number is positive:', numb)
```

## 2. If else Statements

```python
age = 18

if(age >= 18):
    print("You are eligible to vote:")
else:
    print("You are not eligible to vote:")
```

```python
numb = int(input('Enter the number:'))
if(numb > 0):
    print('Number is positive:', numb)
else:
    print('Number is negative:', numb)
```

## 3. elif statements

```python
age = 18

if(age >= 18):
    print("You are a teenager:")
elif(age <= 10):
    print("You are a minor:")
else:
    print('You are not eligible to vote:')
```

```python
numb = int(input('Enter the number:'))
if(numb > 0):
    print('Number is positive:', numb)
elif(numb < 0):
    print('Number is negative:', numb)
else:
    print('Number is zero:', numb)
```

## 4. Nested If statements

```python
a = 10
b = 20
c = 30

if((a>b)):
    if((a > c)):
        print(a , 'is the greatest :')
    else:
        print(c,   'is the greatest :')
else:
    if((b>c)):
        print(b, 'is the greatest :')
    else:
        print(c, 'is the greatest :')
```

## Determine if a number is positive and is even or odd or if it is negative

```python
num = int(input("Enter the number:"))

if(num > 0):
    if((num % 2)== 0):
        print(num,"is even")
    else:
        print(num, "is odd")
else:
    print(num, "is negative:")
```

### Determine if a year is a leap year using nested condition statement


```python
year = int(input("Enter the year:"))

if((year%4 == 0)):
    if((year%100 == 0)):
        if((year%400 == 0)):
            print(year, "is a leap year:")
        else:
            print(year, "is not a leap year:")
    else:
        print(year,"is a leap year:")
else:
    print(year,"is not a leap year:")
```

### Assignment

```python
## 1. ## Assignment
## Simple Calculator program

num1 = int(input("Enter the first number:"))
num2 = int(input("Enter the second number:"))
op   = input("Enter the operation(+, -, *, /): ")

### perform the requested operation
if(op == '+'):
    res = num1 + num2
elif(op == '-'):
    res = num1 - num2
elif(op == "*"):
    res = num1 * num2
elif(op == "/"):
    res = num1 / num2
else:
    res = num1 // num2

print('Result is:', res)
```

## 5. Match statements

```python
day = 4

match day:
    case 1:
        print('Sunday')
    case 2:
        print('Monday')
    case 3:
        print('Tuesday')
    case 4:
        print('Wednesday')
    case 5:
        print('Thursday')
    case 6:
        print('Friday')
    case 7:
        print('Saturday')
    case _:
        print('Invalid')
```

```python
days = 4 

match days:
    case 4 | 5:
        print("It is weekday:")
    case 6 | 7:
        print("It is weekend:")
    case _:
        print("Looking forward to the weekend:")
```
