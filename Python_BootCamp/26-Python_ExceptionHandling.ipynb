## Exception Handling

An exception is an error that occurs during the execution of code.

## try 

It allows us to test the block of code containing the error.

## except

It helps in handling the error.

## else

It helps in executing the code if it does not have any errors.

## finally

It helps in executing the code regardless of the result of try-except block.

```python
## Division by zero exception

try:
    a = 10
    b = (a-10)
    res = a/b
except:
    print("Division by Zero exception has occurred.")
else:
    print(res)
finally:
    print("Final exception has been occurred.")
```

```python
## Name Error Exception

try:
    a = 10
    print(b)
except NameError as e:
    print("Name Error Exception has occurred.")
except:
    print("Some other Exception has occurred.")
finally:
    print("Final Exception has occurred.")
```

```python
# Index Out of Bound Exception

arr = [10,20,30,40,50]

try:
    print(arr[10])
except ZeroDivisionError as e:
    print("Zero Division Error has occured.")
except IndexError as e:
    print("Index Error has occured.")
except NameError as e:
    print("Name Error has occured.")
except:
    print("Some other exception has occured.")
finally:
    print("Final Exception has occured.")
```

```python
# Value Error Exception

s = "abc"

try:
    x = int(s)
except ValueError as e:
    print("Value Error has occurred:")
except IndexError as e:
    print("Index Error has occurred:")
except NameError as e:
    print("Name Error has occurred:")
except ZeroDivisionError as e:
    print("Zero Division Error has occurred:")
else:
    print("The code is correct:")
finally:
    print("The code has finally been executed.")
```

## Raise an exception

As a Python developer you can choose to throw an exception if a condition occurs

# Raise a Value Error if age is negative
''' 
age = -18

if(age < 0):
    raise ValueError("Ages cannot be negative.")
'''

```python
# Raise Value Error saying Ages cannot be negative

try:
    age=-18
    if(age < 0):
        raise ValueError("Ages cannot be negative.")
except ValueError as e:
    print("Value Error has occured.")
except IndexError as e:
    print("Index Error has occured.")
except NameError as e:
    print("Name Error has occured.")
except ZeroDivisionError as e:
    print("Zero Division Error has occured.")
else:
    print("The code is ok.")
finally:
    print("The code has finally been executed.")
```

```python
# Raise Type Error saying that Only Integers are allowed.

s = "abc"

try:
    if not type(s) is int:
        raise TypeError('Only integers are allowed.')
except TypeError as e:
    print('Type Error has occured.')
except NameError as e:
    print('Name Error has occured.')
except IndexError as e:
    print('Index Error has occured.')
except ZeroDivisionError as e:
    print('ZeroDivisionError has occured.')
else:
    print("The code is ok.")
finally:
    print("The code has finally been executed.")
```
