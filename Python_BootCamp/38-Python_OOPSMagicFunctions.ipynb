## MAGIC FUNCTIONS

It refers to the list of methods associated with the entity. It starts with __ end ends with __.

```python
# list of all the methods associated with int


print(dir(int))
```

```python
a = 10

### Adding 10 to a

print(a.__add__(10))

print(a.__mul__(8))
```

```python
# list of all the methods associated with the string

print(dir(str))
```

```python
### Default Constructor to initialize the data members of the class

class Rectangle:
    def __init__(self):
        self.length = 120
        self.breadth = 80

    def print_details(self):
        print("Length of the rectangle is:", self.length)
        print("Breadth of the rectangle is:", self.breadth)

    def calc_area_rec(self):
        return(self.length * self.breadth)

rec = Rectangle()
rec.print_details()
print("Area of the rectabgle is:", rec.calc_area_rec())
```

```python
## list of all the methods associated with the tuple

print(dir(tuple))
```

```python
# length of the tuple

tuple = (10,20,30,40)

print("Length of the tuple is:", tuple.__len__())
```

```python
class person:
    def __init__(self,n,a,s):
        self.name = n
        self.age  = a
        self.salary = s

    def __str__(self):
        return "This is my new object that will return something"

    def print_details(self):
        print(self.name)
        print(self.age)
        print(self.salary)


p = person('parag',30,52345)
p.print_details()

print(p)
```
