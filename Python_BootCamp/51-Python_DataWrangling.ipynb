# Data Wrangling

Data wrangling involves processing the data in various formats like - merging, grouping, concatenating etc. for the purpose of analysing or getting them ready to be used with another set of data

## Merging Data

The Pandas library in python provides a single function, merge, as the entry point for all standard database join operations between DataFrame objects

```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
%matplotlib inline

left = pd.DataFrame({
         'id':[1,2,3,4,5],
         'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
         'subject_id':['sub1','sub2','sub4','sub6','sub5']})

print("---------------------------------------First Dataframe----------------------------------------")
print(left)


print("---------------------------------------Second Dataframe----------------------------------------")
right = pd.DataFrame(
         {'id':[1,2,3,4,5],
         'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
         'subject_id':['sub2','sub4','sub3','sub6','sub5']})

print(right)


print("---------------------------------------Merge two Dataframe using Inner Join----------------------------------------")
pd.merge(left,right,on='id',how='inner')
```

## Grouping Data

Grouping data sets is a frequent need in data analysis where we need the result in terms of various groups present in the data set. Panadas has in-built methods which can roll the data into various groups

```python
# import the pandas library
import pandas as pd

ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
         'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
         'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
         'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
         'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)
print(df)

```

```python
### perform the grouping based on Year

x=df.groupby('Year')
print(x)
```

```python
### Get all the information based on Year '2014'

x.get_group(2014)
```

```python
### Get all the information based on Year '2015'

x.get_group(2015)
```

```python
### Get all the information based on Year '2016'

x.get_group(2016)
```

```python
### Get all the information based on Year '2017'

x.get_group(2017)
```

## Concatenating Data

Pandas provides various facilities for easily combining together Series, DataFrame, and Panel objects. In the below example the concat function performs concatenation operations along an axis. Let us create different objects and do concatenation

```python
import pandas as pd
one = pd.DataFrame({
         'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
         'subject_id':['sub1','sub2','sub4','sub6','sub5'],
         'Marks_scored':[98,90,87,69,78]},
         index=[1,2,3,4,5])
two = pd.DataFrame({
         'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
         'subject_id':['sub2','sub4','sub3','sub6','sub5'],
         'Marks_scored':[89,80,79,97,88]},
         index=[1,2,3,4,5])
```

```python
print(one)
```

```python
print(two)
```

```python
### Concat two dataframes

pd.concat([one,two])
```
