## Inheritance:

It refers to the property of acquiring the properties of the child class from the parent class.

## 1. Single Level Inheritance :

It refers to the process of acquiring the properties of child class from the parent class directly.

```python
class Parent:
    def print_parent_details(self):
        print('Parent Class Details')

class Child(Parent):
    def print_child_details(self):
        print('Child Class Details')


## creating an object for the class
ch = Child()
ch.print_child_details()
ch.print_parent_details()
```

```python
# Addition of two numbers using Single level Inheritance

class Number:
    def enter_two_numbers(self,x,y):
        self.num1 = x
        self.num2 = y

    def print_two_numbers(self):
        print("First Number is:", self.num1)
        print("Second Number is:", self.num2)

class Addition(Number):
    def calc_add_numbers(self):
        print("Addition of two numbers is:", (self.num1 + self.num2))

a = Addition()
a.enter_two_numbers(10,10)
a.print_two_numbers()
a.calc_add_numbers()
```

## 2. Multi Level Inheritance

It refers to the process of acquiring the properties of one class into another class.

```python
class Num1:
    def enter_first_number(self,x):
        self.num1 = x

    def print_first_number(self):
        print("First Number is:", self.num1)

class Num2(Num1):
    def enter_second_number(self,y):
        self.num2 = y

    def print_second_number(self):
        print("Second Number is:", self.num2)

class Total(Num2):
    def print_total(self):
        print("Sum of two numbers is:", (self.num1 + self.num2))


### creating the object for the class
t = Total()
t.enter_first_number(10)
t.print_first_number()

t.enter_second_number(10)
t.print_second_number()

t.print_total()
```

## 3. Multiple Inheritance:


It refers to the process when the class inherits from more than one parent class.

```python
class Num1:
    def enter_first_number(self,x):
        self.num1 = x

    def print_first_number(self):
        print("First Number is:", self.num1)


class Num2:
    def enter_second_number(self,y):
        self.num2 = y

    def print_second_number(self):
        print("Second Number is:", self.num2)


class Total(Num1, Num2):
    def calculate_total(self):
        print("Sum of two numbers is:", (self.num1 + self.num2))


### define the object for the clss
obj = Total()
obj.enter_first_number(10)
obj.print_first_number()


obj.enter_second_number(10)
obj.print_second_number()

obj.calculate_total()
```

## 4. Hierarchical Inheritance

It refers to the process when multiple number of classes gets inherited from the same base class.

```python
class Numbers:
    def enter_numbers(self,x,y):
        self.num1 = x
        self.num2 = y

    def print_numbers(self):
        print("First Number is:", self.num1)
        print("Second Number is:", self.num2)

class Add(Numbers):
    def calc_sum_numbers(self):
        print("Sum of two numbers is :", (self.num1 + self.num2))

class Sub(Numbers):
    def calc_diff_numbers(self):
        print("Difference between two numbers is :", (self.num1 - self.num2))


class Mul(Numbers):
    def calc_product_numbers(self):
        print("Product of two numbers is :", (self.num1 * self.num2))

class Div(Numbers):
    def calc_division_numbers(self):
        print("Diviision of two numbers is :", (self.num1 / self.num2))


a = Add()
a.enter_numbers(10,2)
a.print_numbers()
a.calc_sum_numbers()


s = Sub()
s.enter_numbers(10,2)
s.print_numbers()
s.calc_diff_numbers()


m = Mul()
m.enter_numbers(10,2)
m.print_numbers()
m.calc_product_numbers()


d = Div()
d.enter_numbers(10,2)
d.print_numbers()
d.calc_division_numbers()
```

## super()

It is used to inherit the constructor of the parent class to initailize all the member variables.

```python
class Car:
    ### initialize the constructor to initialize the data members of the class
    def __init__(self,windows,doors,enginetype):
        self.windows = windows
        self.doors   = doors
        self.enginetype = enginetype

    def print_details(self):
        print("Windows    used in the car is:", + self.windows)
        print("doors      used in the car is:",   + self.doors)
        print("enginetype used in the car is:", + self.enginetype)


class Audi(Car):
    ### use the init contructor to inherit the properties from the car class along with horsepower
    def __init__(self,windows,doors,enginetype,horsepower):
        super().__init__(windows,doors,enginetype)
        self.horsepower = horsepower


    def print_details(self):
        print("Windows    used in the car is:", + self.windows)
        print("doors      used in the car is:",   + self.doors)
        print("enginetype used in the car is:", + self.enginetype)
        print("horse power used in the Car is:", + self.horsepower)


a = Audi(4,5,"Diesel", 200)
print(a.windows)
print(a.doors)
print(a.enginetype)
print(a.horsepower)
```
