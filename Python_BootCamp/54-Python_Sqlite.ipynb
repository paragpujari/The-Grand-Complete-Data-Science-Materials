##  SQL And SQLite


SQL (Structured Query Language) is a standard language for managing and manipulating relational databases. SQLite is a self-contained, serverless, and zero-configuration database engine that is widely used for embedded database systems. In this lesson, we will cover the basics of SQL and SQLite, including creating databases, tables, and performing various SQL operations.

```python
import sqlite3
```

```python
## create a connect that is used to connect to the sqlite database

connection1 = sqlite3.connect('exampless.db')
```

```python
## create a cursor object

cursor1 = connection1.cursor()
```

```python
## Create a Table

cursor1.execute(
    '''
    Create Table If Not Exists employees(
    id Integer Primary Key,
    name Text Not Null,
    age Integer,
    department text
    )
    '''
)
```

```python
### fetch the data from the table using the cursor

cursor1.execute(
    '''
     Select * from employees;
    '''
)
```

```python
## Insert the data in sqlite table
cursor1.execute(
    '''
     Insert Into employees(name,age,department)VALUES('Krish',32,'Data Scientist')
    '''
)


cursor1.execute(
    '''
     Insert Into employees(name,age,department)VALUES('Bob', 25, 'Engineering')
    '''
)


cursor1.execute(
    '''
     Insert Into employees(name,age,department)VALUES('Charlie', 35, 'Finance')
    '''
)
```

```python
### fetch the data from the table using the cursor

cursor1.execute(
    '''
     Select * from employees;
    '''
)

rows = cursor1.fetchall()

for row in rows:
    print(row)
```
