## Python Dates

It has a module datetime to work with the dates.

```python
import datetime

### get the current datetime

x = datetime.datetime.now()

print(x)
```

```python
# Return the year and name of weekday

### year

print('Year is:', x.year)

### name of weekday

print('Name of weekday is:', x.strftime("%A"))
```

```python
# Abbrebriate weekday name

print("Abbrebriated Weekday name is:", x.strftime("%a"))
```

## Creating Date Objects

To craete a date object, we need to use  a function named datetime that accepts three parameters year, month and day.

```python
import datetime

x = datetime.datetime(2025,12,26)

print(x)
```

## The strftime() Method

It stands for string format time that is used to convert the datetime object into readable strings.

It accepts one parameter i.e format(format of the returned string.)

```python
# create a datetime object

import datetime

x = datetime.datetime(2025,12,26)

print(x)

### Week day

print("Weekday is :", x.strftime("%A"))
```

```python
# Return the full month name of the string

print("Full Month string is :", x.strftime("%B"))
```
