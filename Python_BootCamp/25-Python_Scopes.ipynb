## Python Scope

A variable is available inside the region where it is created. This is called region.

## Local Scope

It is the region where the variable is created inside the particular function.

```python
# Local scope

def myfunccall():
    x = 200  ### x is a local variable as it is used inside the function. It is a local scope.
    print(x)
    
### function call
myfunccall()
```

## Function Inside Function

When one function is used inside another function, then it is function inside function.

```python
### function inside function

#### function definition
def myfunc():
    x = 100
    def myinnerfunc():
        x = 200
        print('Value used inside the inner function is:', x)
    myinnerfunc()
    print('Value used inside the outside function is:', x)
myfunc()
```

## Global Scope

When the variable is used outside the function block then it is said to be in global scope.

```python
x = 60
def myfunc():
    x = 80
    def innermyfunc():
        x = 100
        print('Value used in the inner local scope is:', x)
    innermyfunc()
    print('Value used in the local scope is:', x)
myfunc()
print('Value used in the global scope is:', x)
myfunc()
```

## Global Keyword

If we have a variable declared with a Global keyword inside the function, then it can be accessible inside as well as outside the function.

```python
# Declare a variable 'x' as global inside the function

# function definition

def myfunc():
    global x
    x = 200
    print('Global variable used inside the function is:', x)
myfunc()

print('Variable used outside the function is:', x)
```

```python
# Declare a variable 'x' as global inside the function
x1 = 200

def myfunc():
    global x
    x = 150
    print('Global Variable used inside the function is :', x)
myfunc()

print('Global Variable used outside the function is:', x)

print('Variable used for outer scope is:', x1)
```

## Nonlocal Keyword

If we declare a variable as a Nonlocal keyword in a nested function, then it can be accessible to the outer functions.

```python
### function definition
def myfunccall():
    x = 200
    def myinnerfunccall():
        nonlocal x
        x = 300
    myinnerfunccall()
    return(x)


print(myfunccall())
```

```python
### function definition
def myfunccall():
    x = 200
    def myinnerfunccall():
        nonlocal x
        x = 300
    return(x)
    myinnerfunccall()


print(myfunccall())
```
