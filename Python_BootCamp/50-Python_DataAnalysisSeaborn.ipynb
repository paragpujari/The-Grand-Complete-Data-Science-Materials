## Plotting a Displot

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x = np.array([1,2,3,4,5])

sns.distplot(x,kde=True)
```

```python
## Customize the scatter plot

import seaborn as sns 
import matplotlib.pyplot as plt


df = sns.load_dataset('tips')

### plot the scatter plot between the total-bill vs tip

sns.scatterplot(x='total_bill',y='tip',data=df)

# give a title to the graph
plt.title("Relationship between total bill and tip")

# give a name to the x label
plt.xlabel("Total Bill")

# give a name to the y label
plt.ylabel("Tip")

# plot the graph
plt.show()
```

## OBSERVATION

(i) From the above scatter plot distribution, we can see that the male are more in number than the female who have paid the bill.

```python
## Peform the Histogram for the Petal length in iris dataset

import seaborn as sns 
import matplotlib.pyplot as plt 

# load the dataset
df = sns.load_dataset('iris')


## Peform the Histogram for the Petal length
sns.histplot(data=df,x='petal_length',kde=True)

## Add the title and labels to the graph
plt.title("Distribution of petal length in Iris Flowers")

plt.xlabel("Petal Length")

plt.ylabel("Frequency")

plt.show()
```

### OBSERVATIONS:

1. The above graph shows the distributions of the entire dataset with respect to the 'petal_length' with the smoothening curve given infront of it.

```python
## perform the density plot for the total bill vs Meal time

import seaborn as sns 
import matplotlib.pyplot as plt 

# load the dataset
df = sns.load_dataset("tips")


# plot the smoothened graph for the total bill vs Meal time
sns.kdeplot(data=df,x="total_bill",hue="time",fill=True,alpha=0.6,linewidth=1.5)

# Add a title and labels to the plot using Matplotlib

plt.title("Density Plot of Total Bill by Meal Time")

plt.xlabel("Total Bill ($)")

plt.ylabel("Density")




# Show the plot

plt.show()
```

### OBSERVATIONS:

1. Maximum number of customers are having dinner than lunch.

```python
## perform a box plot for day vs total bill based on time


sns.boxplot(x="day",y="total_bill",data=df,hue="time")
```

### OBSERVATIONS:

1. People are having dinner more in weekends than the people having lunch in the weekdays.

## Seaborn heatmaps

A heatmap is a graphical representation of data that uses colors to depict the value of a variable in a two-dimensional space. Heatmaps are commonly used to visualize the correlation between different variables in a dataset

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Load the dataset

tips = sns.load_dataset('tips')
## print(tips)

x = tips[['total_bill','tip','size']]

print(x)
```

```python
dfa = x.corr()

print(dfa)
```

```python
sns.heatmap(dfa,annot=True)
```
