## Python Identity Operators

Identity operators are used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location:

```python
x = ["apple","banana"]
y = ["apple","banana"]
z = x     #### point to the same memory address

print(x)
print(y)
print(z)


print(id(x))
print(id(y))
print(id(z))



print(x is y)

print(x is z)
```

## Python Membership Operators

Membership operators are used to test if a sequence is presented in an object:

```python
res = [10,20,30,40,50]


print(10 in res)
```

```python
x = ["apple", "banana"]


print("cherry" not in x)
```

## Python Bitwise Operators

Bitwise operators are used to compare (binary) numbers

```python
print(6 & 3) # Bitwise AND
```

```python
print(6 | 3) # Bitwise OR
```

```python
print(6 ^ 3) # Bitwise XOR
```

```python
print(3 << 2) # Bitwise AND  # Shits the specific number of bits to the left.
```

```python
print(8 >> 2) # Bitwise AND  # Shits the specific number of bits to the left.
```
