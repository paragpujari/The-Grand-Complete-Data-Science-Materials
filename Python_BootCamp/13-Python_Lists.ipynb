## Lists

Lists are used to store multiple items in a single variable.

Lists are one of 4 built-in data types in Python used to store collections of data, the other 3 are Tuple, Set, and Dictionary, all with different qualities and usage.


Lists are created using square brackets:

```python
### empty lists

lists = []

print(lists)

print(type(lists))
```

```python
## lists with elements

lists = ['apple','banana','cherry','guava']

print(lists)

print(type(lists))
```

## Properties of Lists

1. Lists are ordered, mutuable and allow duplicates.

```python
# Lists allow duplicate values:


lists = ['apple','banana','banana','cherry','cherry','guava']

print(lists)

print(type(lists))
```

2.  Lists determine the total number of elements using len() function.

```python
print(lists)

x = len(lists)

print(x)
```

3. Lists can contain the elements of the same or different types.

```python
lists = [10,20,30,3.456,23.456,8+9j,"parag",(30,50),[9,3,4],True,"apple","male",False]

print(lists)

print(type(lists))
```

4. type() -----> It defines the type of the list.

```python
print(lists)


print(type(lists))
```

5. list() ---------------> It is a constructor that helps in creating a new list.

```python
mylist = list(('apple','banana','guava'))

print(mylist)

print(type(mylist))
```

## Python - Access Lists Items

Access Items

List items are indexed and you can access them by referring to the index number:

```python
### print the first element in the lists

thislist = ["apple", "banana", "cherry"]

thislist[0]
```

```python
### print the second element in the lists

thislist = ["apple", "banana", "cherry"]

thislist[1]
```

## Negative Indexing

Negative indexing means start from the end

-1 refers to the last item, -2 refers to the second last item etc.

```python
### print the last item in the lists

thislist = ["apple", "banana", "cherry"]

print(thislist[-1])
```

```python
### Print the second last item in the list

print(thislist[-2])
```

## Range of Indexes

You can specify a range of indexes by specifying where to start and where to end the range.

When specifying a range, the return value will be a new list with the specified items.

```python
# Return the third fourth and fifth item in the lists


thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]

thislist[2:5]
```

```python
# Return all the elements before kiwi

thislist[0:4]
```

```python
# This example returns the items from "cherry" to the end

thislist[2:]
```

## Range of Negative Indexes

Specify negative indexes if you want to start the search from the end of the list:

```python
# This example returns the items from "orange" (-4) to, but NOT including "mango" (-1):

thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]

thislist[-4:-1]
```

## Sort List Alphanumerically

List objects have a sort() method that will sort the list alphanumerically, ascending, by default

```python
# Sort the list alphabetically

thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]

thislist.sort()

thislist
```

```python
# Sort the list numerically

thislist = [100, 50, 65, 82, 23]

thislist.sort()

thislist
```

## Sort Descending

To sort descending, use the keyword argument reverse = True

```python
thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]

thislist.sort(reverse=True)

thislist
```

```python
thislist = [100, 50, 65, 82, 23]

thislist.sort(reverse=True)

thislist
```

## Case Insensitive Sort

By default the sort() method is case sensitive, resulting in all capital letters being sorted before lower case letters

```python
thislist = ["banana", "Orange", "Kiwi", "cherry"]

thislist.sort()

thislist
```

```python
# Perform a case-insensitive sort of the list

thislist = ["banana", "Orange", "Kiwi", "cherry"]

thislist.sort(key=str.lower)

thislist
```

## Using a While Loop

You can loop through the list items by using a while loop.

Use the len() function to determine the length of the list, then start at 0 and loop your way through the list items by referring to their indexes.

Remember to increase the index by 1 after each iteration.

```python
## Print all items, using a while loop to go through all the index numbers

thislist = ["apple", "banana", "cherry"]

l = len(thislist)

i = 0

while(i < l):
    print(i, thislist[i])
    i = i + 1
```

## List Comprehension

List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.

Example:

Based on a list of fruits, you want a new list, containing only the fruits with the letter "a" in the name.

Without list comprehension you will have to write a for statement with a conditional test inside

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]


res = [x for x in fruits if(x != "")]

res
```

```python
# Only accept items that are not "apple"

fruits = ["apple", "banana", "cherry", "kiwi", "mango"]


res = [x for x in fruits if(x != "apple")]

res
```

```python
# You can use the range() function to create an iterable 

res = [x for x in range(11) if(x != 0)]

res
```

```python
# Accept only numbers lower than 5

res = [x for x in range(10) if(x < 5)]

res
```

```python
# Set the values in the new list to upper case


fruits = ["apple", "banana", "cherry", "kiwi", "mango"]


res = [x.upper() for x in fruits if(x != "")]

res
```

```python
# Set all values in the new list to 'hello'

res = ["hello" for x in fruits if(x != "")]

res
```

```python
# Return "orange" instead of "banana"


fruits = ["apple", "banana", "cherry", "kiwi", "mango"]


res = ["orange" if(x == "banana") else x for x in fruits]
res
```
