```python
# Print the areas and perimeters of Rectangle, Squares, Circles and Triangles

## define the abstract class
from abc import ABC, abstractmethod

### define the abstract method
class Shape(ABC):
    def print_area(self):
        pass

### define the non abstract method
class Rectangle(Shape):
    def enter_rec_details(self,l,b):
        self.length = l
        self.breadth = b

    def calc_rec_area(self):
        print('Area of the rectangle is:', (self.length * self.breadth))

    def calc_rec_perimeter(self):
        print('Perimeter of the rectangle is:', 2*(self.length + self.breadth))


class Square(Shape):
    def enter_square_details(self,s):
        self.side = s

    def calc_square_area(self):
        print('Area of the square is:', (self.side * self.side))

    def calc_square_perimeter(self):
        print('Perimeter of the square is:', 4*(self.side))


class Circle(Shape):
    def enter_circle_details(self,r):
        self.radius = r

    def calc_circle_area(self):
        print('Area of the circle is:', 3.142 * (self.radius * self.radius))

    def calc_circle_perimeter(self):
        print('Perimeter of the circle is:', 2 * 3.142 * (self.radius))


class Triangle(Shape):
    def enter_triangle_details(self,ba,ht):
        self.base = ba
        self.height = ht

    def calc_triangle_area(self):
        print('Area of the triangle is:', 0.5 * (self.base * self.height))

    def calc_triangle_perimeter(self):
        print('Perimeter of the triangle is:', (self.base + self.height))


print('---------------------------------------------------------------------------------------------------------------------------------------------')

rec = Rectangle()
rec.enter_rec_details(34,43)
rec.calc_rec_area()
rec.calc_rec_perimeter()

print('---------------------------------------------------------------------------------------------------------------------------------------------')

sq = Square()
sq.enter_square_details(40)
sq.calc_square_area()
sq.calc_square_perimeter()

print('---------------------------------------------------------------------------------------------------------------------------------------------')

cir = Circle()
cir.enter_circle_details(20)
cir.calc_circle_area()
cir.calc_circle_perimeter()

print('---------------------------------------------------------------------------------------------------------------------------------------------')

tr = Triangle()
tr.enter_triangle_details(89,64)
tr.calc_triangle_area()
tr.calc_triangle_perimeter()

print('---------------------------------------------------------------------------------------------------------------------------------------------')
```
