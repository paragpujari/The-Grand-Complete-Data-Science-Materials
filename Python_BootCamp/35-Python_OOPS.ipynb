##  1. Perform the addition of two numbers?

```python
### enter the two numbers

a = int(input('Enter the first number:'))
b = int(input('Enter the second number:'))

class numbers:
    def enter_two_numbers(self,a,b):
        self.x = a
        self.y = b

    def print_two_numbers(self):
        print("First Number is:", self.x)
        print("Second Number is:", self.y)

    def calc_add_two_numbers(self):
        self.sum = self.x + self.y
        
    def print_result(self):
        print("Addition of two numbers is:", self.sum)


obj = numbers()
obj.enter_two_numbers(a,b)
obj.print_two_numbers()
obj.calc_add_two_numbers()
obj.print_result()
```

##  2. Perform the multiplication of two numbers?

```python
### enter the two numbers

a = int(input('Enter the first number:'))
b = int(input('Enter the second number:'))


class Numbers:
    def enter_two_numbers(self,a,b):
        self.x = a
        self.y = b

    def print_two_numbers(self):
        print("First Number is:", self.x)
        print("Second Number is:", self.y)

    def multiply_two_numbers(self):
        return(self.x * self.y)

obj = Numbers()
obj.enter_two_numbers(10,10)
obj.print_two_numbers()
print("Multiplication of two numbers is:", obj.multiply_two_numbers())
```
