## Python Generator:

It is like a function that does not store any value but returns the value one after the another using yield keyword.

It is useful for sequences like Fibonacci series the values can grow indefinitely by returning the current state one after the another.

```python
# Use the generator to generate the Fibonacci series

def fibonacci(n):
    a,b = 0,1
    for _ in range(n):
        yield a
        a,b = b,a+b

        
n = int(input("Enter the number:"))

for i in fibonacci(n):
    print(i)
```

```python
### Get the Squares of the numbers

def squares(n):
    for i in range(1,(n+1)):
        yield(i**2)

n = int(input('Enter the number:'))

for i in squares(n):
    print(i)
```
