```python
# Define the class "Employee"

class Employee:
    ### enter the details of the employee
    def enter_emp_details(self,n,a,s):
        self.name = n
        self.age  = a
        self.salary = s

    ### print the employee details
    def print_emp_details(self):
        print("Name is:", self.name)
        print("Age is:", self.age)
        print("Salary is:", self.salary)

    ### define the static method
    @staticmethod
    def print_mentor(mentor):
        print(mentor)

    ### define the another static method
    @staticmethod
    def print_student_list(student):
        print(student)



emp = Employee()
emp.enter_emp_details("Parag",30,67345)
emp.print_emp_details()
Employee.print_mentor("Akhilesh")
Employee.print_student_list("Parag, Nishant, Subham")
```
