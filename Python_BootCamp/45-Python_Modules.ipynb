## Module

It refers to the set of functions that we want to include in our application.

```python
### predefined modeules

import math

x = math.sqrt(16)

print(x)
```

```python
### user defined modules

import add

x = add.addition(4,5)

x
```

```python
import sub

x = sub.substraction(9,4)

x
```

```python
import mul

x = mul.multiply(10,4)

x
```

```python
import divide

x = divide.division(10,5)

x
```

```python
import square

x = square.squares(4)

x
```
