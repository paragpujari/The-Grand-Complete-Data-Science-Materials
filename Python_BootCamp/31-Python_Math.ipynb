## Math Module

It is an inbuilt module used in Python that is used to perform mathematical operations in Python.

###  1. sqrt() = It returns the square root of the number.

```python
import math

x = math.sqrt(16)

print(x)
```

### 2. ceil() = It is used to round off the number upwards to its nearest integer.

```python
import math

x = math.ceil(1.2)

print(x)
```

### 3. floor() = It is used to round off the number downwards to its nearest integer.

```python
import math

x = math.floor(1.2)

print(x)
```

### 4. pi = It the value of PI (3.14...)

```python
import math

print(math.pi)
```
