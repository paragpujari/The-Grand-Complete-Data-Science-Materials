## Comments

It explains the meaning of each and every line of the code.

It makes the code more readable and more understandable.

It prevents the python code from getting executed when testing the code.


## Types of Comment

1. Single line comment

```python
# This is the comment 

print("Hello, World !")
```

2. Multi Line Comment

```python
# This is the comment 
# written in one line
# by Parag

print("This is my favorite class")
```

3. Multiline Comment

```python
# adds the multi line string within triple quotes.

"""
This is the comment
written
in more than one line.
"""

print("Hello, World !")
```
