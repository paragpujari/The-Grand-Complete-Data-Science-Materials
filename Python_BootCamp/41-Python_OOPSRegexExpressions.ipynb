## Regex

Regex stands for Regular Expressions

It is a tool for matching and handling string in python.

It is used to work with the regular expressions.

It also uses an inbuilt module 're' for working with the regular expressions.


## search()

It is used to check for the pattern in the text.

```python
import re

# Check if the word is present in the text

text = "The Bodyguard is the best album."

pattern = r"Body"

### check if the particular pattern is found in the text
res = re.search(pattern,text)

print(res)

if(res):
    print("Match found :")
else:
    print("Match not found: ")
```

\d -------------> It matches for any digit in the text.

```python
import re

# check if any number is present in the text or not

text = "My phone number is 7978304194"

pattern = r"\d"

res = re.search(pattern,text)

if(res):
    print("Match found:")
else:
    print("Match not found: ")
```

### 3. findall() = It finds all occurrences of a specified pattern within a string

```python
import re

s2 = "The BodyGuard is the best album of 'Whitney Houston'."

# Using findall() find out all the occurrances of st in the string

pattern = r'st'

res = re.search(pattern,s2)

print(res)

if(res):
    print("Match found:")
else:
    print("Match not found: ")
```

### 4. split() = It splits a string into an array of substrings based on a specified pattern.

r"\s" --> It  is a regular expression pattern that matches any whitespace characters in the string.

```python
import re

s2 = "The BodyGuard is the best album of 'Whitney Houston'."

pattern = r"\s"

res = re.split(pattern,s2)

print(res)
```

```python
## Search the string to see if it starts with "The" and ends with "Spain"

import re

txt = "The rain in Spain"

res = re.search('^The.*Spain$',txt)

print(res)

if(res):
    print("Match found:")
else:
    print("Match not found: ")

```

### sub() = It replaces the match with the text of your choice.

```python
## Replace every white space character with the number '9'

import re

text = "The rain in Spain"

res = re.sub("\s","9",text)

print(res)
```
