## Python Iterators

An iterator is an object that contains countable number of values.

It can be iterated upon meaning we can traverse through all the values.

## Two methods used in Iterators:

1.   iter() ---> It is used to generate an iterator.

2.   next() ---> It is used to generate the next element .

```python
# Return an iterator from a tuple, and print each value

# Return an iterator from the tuple

fruits = ("apple","banana","cherry","guava","lichi")

print(fruits)

x = iter(fruits)

print(x)
```

### OBSERVATIONS:

1. An iter is an iterator object that helps in iterating through all the elements in the tuples and returns the value.

```python
# print each value from the iterable object

print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
```

```python
## Lists are iterable and we can iterate through the lists using iterator

lists = [10,20,30,40,50]

# create an iterable object(Iterator) from the lists

x = iter(lists)

print(x)


# print each value from the iterable object

print(next(x))
print(next(x))
print(next(x))
print(next(x))
print(next(x))
```

```python
# Strings are also iterable objects, containing a sequence of characters

x = "banana"

for i in x:
    print(i)
```

```python
# Iterator the string through the iterator

myiter = iter(x)
myiter


print(next(myiter))
print(next(myiter))
print(next(myiter))
print(next(myiter))
print(next(myiter))
print(next(myiter))
```
