## Decorators

1.   Decorators let you add extra behavior to a function, without changing    the function's code.

2.   A decorator is a function that takes another function as input and returns a new function


3.   Define the decorator first, then apply it with @decorator_name above the function

```python
# define the decorator function to convert the text "hello parag" to upper case
def mydecorator(myfunctiondata):
    def inner():
        return(myfunctiondata().upper())
    return(inner)


### define the decorator function
@mydecorator
def myfunctiondata():
    return("hello parag")

### function call
print(myfunction())
```

# Multiple Decorator Calls

A decorator can be called multiple times. Just place the decorator above the function you want to decorate

```python
### convert all the texts into upper case
def changecase(func):
    def inner():
        return(func().upper())
    return(inner)
    
### decorator
@changecase
def myfirstcase():
    return("hello parag")

@changecase
def mysecondcase():
    return("I am sorry")

@changecase
def mythirdcase():
    return("I live in delhi")

### function call
print(myfirstcase())
print(mysecondcase())
print(mythirdcase())
```

## Arguments in the Decorated Function

Functions that requires arguments can also be decorated, just make sure you pass the arguments to the wrapper function

```python
# Here we are passing the arguments in the decorator

# define the decorator function to convert the text into upper case.

def mydecorator(myfunctioncall):
    def inner(x):
        return(myfunctioncall(x).upper())
    return(inner)

### call the decorator function
@mydecorator
def myfunctioncall(name):
    return('Hello ' + ' ' + name)

#### function call
print(myfunctioncall("Parag"))
```
