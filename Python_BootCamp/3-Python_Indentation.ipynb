## Indentation

Indentation refers to the spaces at the beginning of the code line.

It is used to indicate the block of code.

```python
if(5 > 2):
    print("5 is greater than 2:")
```

```python
if(5 >= 2):
    print("5 is greater than or equals to 2:")
elif(5 == 2):
    print("5 is equals to 2:")
else:
    print("5 is less than 2:")
```
