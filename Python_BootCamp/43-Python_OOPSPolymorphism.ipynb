## Polymorphism:

It refers to the ability of an object to exists in multiple forms.

```python
class Test:
    def add(self,x,y):
        print(x+y)


t = Test()
t.add(10,20)
t.add('Parag','Pujari')
t.add([10,20,30],[40,50,60])
t.add((10,20,30),(40,50,60))
```

## Types of Polymorphism:

1. Method Overriding

It refers to the ability of child class to inherit some of the features of the parent class.

```python
class Animal:
    def eating(self):
        print("The Animal is hungry.")

class Dog(Animal):
    def eating(self):
        print("The dog is barking as he is hungry.")

d = Dog()
d.eating()
```

2. Method Overloading

It refers to the process of using the same function name but with different number of arguements.

```python
class Poly:
    def add(self,x,y):
        print(x + y)

p = Poly()
p.add(10,20)
p.add("Parag","Pujari")
p.add([10,20,30],[40,50,60])
```
