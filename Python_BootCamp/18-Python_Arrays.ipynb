## Arrays

Arrays are used to store multiple values in one single variable.

```python
# array

## Create an array containing car names

cars = ["Ford","Maruti","BMW"]

print(cars)

print(type(cars))
```

## Access the Elements of an Array

You refer to an array element by referring to the index number

```python
## Get the value of the first array item
cars[0]
```

```python
## Get the value of the last array item
cars[-1]
```

```python
## Modify the value of the first array item

cars[0] = "Toyota"

print(cars)
```

## The Length of an Array

Use the len() method to return the length of an array (the number of elements in an array)

```python
print('Length of the array is:', len(cars))
```

## Looping Array Elements

You can use the for in loop to loop through all the elements of an array

```python
for i in cars:
    print(i)
```

## Adding Array Elements

You can use the append() method to add an element to an array

```python
cars.append('Honda')
```

```python
cars
```

## Methods used in Arrays :

```python
# 1. append() = It is used to add the element in the array at the end.

fruits = ['apple', 'banana', 'cherry']
# Add the fruit 'orange' at the end
fruits.append('orange')

print(fruits)
```

```python
# 2. clear() = It removes all the elemnts from the list.


fruits.clear()

print(fruits)
```

```python
# 3. copy() = It returns the copy of the list.

fruits = ['apple', 'banana', 'cherry', 'orange']

print(fruits)

x = fruits.copy()

print(x)
```

```python
fruits[1] = 'guava'


print(fruits)

print(x)
```

## Observation:

(i) Using copy(), a shallow copy is created i.e any changes happened in the original array if not reflected in the copied array.

```python
# 4. count() = It returns the total number of elements in the array.

fruits = ['apple', 'banana', 'cherry']


x = fruits.count('cherry')

print(x)
```

```python
# Return the number of times the value 9 appears int the list

points = [1, 4, 2, 9, 7, 8, 9, 3, 1]


x = points.count(9)

print(x)
```

```python
# 5. extend() =  It adds all the elements of one array at the end of current array.

fruits = ['apple', 'banana', 'cherry']

cars = ['Ford', 'BMW', 'Volvo']


fruits.extend(cars)

print(fruits)
```

```python
# 6. index() = It returns the position at the first occurrence of the specified value

fruits = [4, 55, 64, 32, 16, 32]

# position of 32

x = fruits.index(32)

print(x)
```

```python
fruits = ['apple', 'banana', 'cherry', 'kiwi', 'mango', 'orange', 'cherry']

# print(fruits.index('cherry'))

# Find the postion of 'cherry', but start the search at position 5


x = fruits.index('cherry',5)

print(x)
```

```python
# 7. insert() =  It inserts the element at the specified index position.

fruits = ['apple', 'banana', 'cherry']

# Insert the value "orange" as the second element of the fruit list

fruits.insert(1,'orange')


print(fruits)
```

```python
# 8. pop() = It is used to remove the element from the array using the index position.

fruits = ['apple', 'banana', 'cherry']

## Remove the second element of the fruit list

fruits.pop(1)


print(fruits)
```

```python
# Removes the last inserted value

fruits.pop()

print(fruits)
```

```python
# 9. remove() = It is used to remove the element from the array using the name.

# Remove the "banana" element of the fruit list

fruits = ['apple', 'banana', 'cherry']

fruits.remove('banana')

print(fruits)
```

```python
# 10. reverse() = It is used to reverse all the elements in the array

fruits = ['apple', 'banana', 'cherry']


fruits.reverse()


print(fruits)
```

```python
# 11. sort() = It arranges all the elements in the array alphabetically in ascending or descending order.


cars = ['Ford', 'BMW', 'Volvo']


print(cars)


cars.sort()

print(cars)


#### Reverse the elements of the array in reverse order


cars.sort(reverse=True)

print(cars)
```
