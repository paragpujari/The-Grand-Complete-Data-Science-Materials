### while Loop

```python
### Numbers from 1 till 10

i = 1
while(i <= 10):
    print(i)
    i = i + 1
```

```python
## Loop Control Statements

## break
## The break statement exits the loop permaturely

## break statement


for i in range(10):
    if(i == 5):
        break
    print(i)
```

```python
### Print prime numbers using break

for i in range(2, 10):
    for x in range(2, i):
        if((i % x) == 0):
            break
    else:
        print(i, end = " ")
```

```python
## continue

## The continue statement skips the current iteration and continues with the next.

### skip all the numbers divisible by 2

for i in range(1,100):
    if((i%2)==0):
        continue
    print(i, end= " ")
```

```python
## pass
## The pass statement is a null operation; it does nothing.

for i in range(5):
    if(i == 3):
        pass
    print(i)
```

```python
### Calculate the sum of first N natural numbers using a while and for loop

N = 10

sum = 0

i = 1

while(i <= N):
    sum = sum + i
    i = i + 1

print("Sum of N Natural Numbers is :", sum)
```

### Nested Loops

```python
## Nested loopss
## a loop inside a loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```
