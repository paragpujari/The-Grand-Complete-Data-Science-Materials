## Python String Formatting

String formatting is a way of changing selected parts of the string.


```python
text = f"the price is 49 dollars."

print(text)
```

## Placeholders and Modifiers

To format values in an f-string, add placeholders {}, a placeholder can contain variables, operations, functions, and modifiers to format the value

```python
price = 49

print(f"The price of the bag is {price} dollars.")
```

```python
## Display the price with 2 decimals

price = 49

print(f"The price of the bag is {price:.2f} dollars.")
```

```python
## Display the value 95 with 2 decimals

price = 95

print(f"The price of the bag is {price:.2f} dollars.")
```

## Perform Operations in F-Strings

You can perform Python operations inside the placeholders.

You can do math operations

```python
## Perform a math operation in the placeholder, and return the result

price = 95

print(f"The price of the bag is {(price*20):.2f} dollars.")
```

```python
## Add taxes before displaying the price

price = 99

tax   = 0.25

print(f"The price of the bag after paying tax is {(price + (price * tax))} dollars.")
```

```python
quantity = 3
itemno = 567
price = 49

text = "I want {} pieces of itemno {} for {:.2f} dollars."

print(text.format(quantity,itemno,price))
```

## Index Numbers


You can use index numbers (a number inside the curly brackets {0}) to be sure the values are placed in the correct placeholders

```python
quantity = 3
itemno = 567
price = 49


print('I want {0} pieces of itemno {1} for {2:.2f} dollars'.format(quantity, itemno,price ))
```

```python
age = 36
name = "John"


print("Hi, my name is {1} and I am {0} years old".format(age,name))
```
