##  Reference Counting

Reference Counting refers to the technique used in Memory Management in Python for memory allocation and deallocation. Each Python object has a reference object allocated to it. When ever a reference object gets incremented then the memory gets allocated to the object else the memory from the particular object gets deallocated from the object.

```python
import sys 

a = []

print(sys.getrefcount(a))
```

## OBSERVATIONS:

1. Memory Management is done using reference counting.

2. The reference count has the memory allocated to the particular Python object. It's count increases when the memory gets allocated to the python object else its count reduces when the memory gets deallocated from the object.

3. By defauly getrefcount has the memory already allocated to the object. So its reference count is 1. Also a(empty list) is added to the getrefcount. So its values becomes 2.

```python
b = a 

print(sys.getrefcount(b))
```

## Garbage Collection

It is an important technique used in Python that helps in removing the memory allocated to the Python object which is non longer required in the program.

### Cyclic Garbage Collector

It is an important technique used in Memory Management in Python. It occurs when the two or more objects references each other and thus prevents their reference count from becoming zero.

So Cyclic Garbage Collector is used to check for all those Python objects that references each other and frees all those memory from the objects that references each other and make their reference count to 0.

```python
# enable the garbage collection

import gc

print(gc.enable())
```

## OBSERVATIONS:

1. It helps in enabling the garbage collection process.

```python
# disable the garbage collection

import gc 

print(gc.disable())
```

```python
# collect all the garbage collection 

print(gc.collect())
```

## OBSERVATIONS:

1. It enables the Cyclic Garbage Collector and finds out the list of all the Python objects that references each other in a circular reference and are alos unreachable.

## Memory Management Best Practices

1.   Use Local Variables: Local variables have a shorter lifespan and are freed sooner than global variables.

2.   Avoid Circular References: Circular references can lead to memory leaks if not properly managed.

3.   Use Generators: Generators produce items one at a time and only keep one item in memory at a time, making them memory efficient.

4.   Explicitly Delete Objects: Use the del statement to delete variables and objects explicitly.

5.   Profile Memory Usage: Use memory profiling tools like tracemalloc and memory_profiler to identify memory leaks and optimize memory usage.

```python
## Handled Circular reference

import gc 

class MyObject:
    ## Memory creation
    def __init__(self,name):
        self.name = name
        print(f"Object {self.name} created")

    ## MEMORY deleted
    def __del__(self):
        print(f"Object {self.name} deleted")

# create the object for the memory class
obj1 = MyObject("obj1")
obj2 = MyObject("obj2")

# objects references each other
obj1.ref = obj2 
obj2.ref = obj1


del obj1 
del obj2 

# trigger the cyclic garbage collector
gc.collect()
```

```python
## Generators For Memory Efficiency
#Generators allow you to produce items one at a time, using memory efficiently by only keeping one item in memory at a time.

def generate_numbers(n):
    for i in range(n):
        yield i

## using the generator
for num in generate_numbers(100000):
    print(num)
    if num>10:
        break
```
