# File Handling

It is one of the most important operation used in Python that is used for creating, reading and deleting the files.

```python
# Write the contents of the file

with open('data1.txt','w') as f:
    f.write('Hi, My name is Parag.\n I am working at Motherson Technologies Services Limited. \n I am a Data Analyst. \n I enjoy learning Python and SQL. \n I want to become a Data Scientist. \n I like working on new projects. \n I am interested in machine learning. \n I spend my free time studying and practicing. \n I live in Noida, Uttar Pradesh. \n I am working hard to improve my career. \n I want to join a big MNC in the future. \n I believe in consistent learning and growth.')
### close the file
f.close()
```

```python
# Read the entire file

with open('data1.txt','r') as f:
    print(f.read())

### close the file
f.close()
```

```python
# Read one line of the file

with open('data1.txt','r') as f:
    print(f.readline())
```

### OBSERVATIONS:

1. readline() -----> It helps in reading one line of the file 

```python
# Read two lines of the file

with open('data1.txt','r') as f:
    print(f.readline())
    print(f.readline())
## close the file
f.close()
```

```python
# Loop through the file line by line

with open('data1.txt','r') as f:
    for i in f:
        print(i)
## close the file
f.close()
```

## Write to an Existing File

To write to an existing file, you must add a parameter to the open() function:

"a" - Append - will append to the end of the file

"w" - Write - will overwrite any existing content

```python
# Open the file "data1.txt" and append content to the file

### append the content into the file

with open('data1.txt','a') as f:
    ### write the content into the file at the end
    f.write('\n I like to solve the problems using datas.')
    

### close the file
f.close()
```

```python
### read the content from the file

with open('data1.txt','r') as f:
    ### read the content from the file
    print(f.read())
    

### close the file
f.close()
```

## Overwrite Existing Content

To overwrite the existing content to the file, use the w parameter

```python
with open('data1.txt','w') as f:
    f.write("Woops! I have deleted the content!")

f.close()
```

```python
### Read the contant from the file

with open('data1.txt','r') as f:
    print(f.read())

f.close()
```

## Delete a File

To delete a file, you must import the OS module, and run its os.remove() function

```python
# Delete the file data1.txt

import os

os.remove("data1.txt")
```

## Check if File exist:

To avoid getting an error, you might want to check if the file exists before you try to delete it

```python
# Check if the file exists , if yes then delete else print the file is already deleted

import os

### check if the path exists

x = os.path.exists("data1.txt")

print(x)

if(x):
    ### delete the file
    os.path.delete("data1.txt")
else:
    print("The file is already deleted")
```
