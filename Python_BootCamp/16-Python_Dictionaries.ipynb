## Dictionaries

(i)   These are the data structures that have key and value pairs enclosed within curly brackets.

(ii)   They  are ordered.

(iii)  They are changable.

(iv)    They do not allow duplicates.(No two keys can have the same value).

```python
# empty dictionary

mydict = {

}


print(mydict)


print(type(mydict))
```

```python
# dictionary with elements

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

print(thisdict)

print(type(thisdict))
```

## Python - Remove Dictionary Items

## Removing Items

There are several methods to remove items from a dictionary

```python
# The pop() = It removes the item with the specified key name

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

print("-----------------Before Updating---------------------")
print(thisdict)


thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

print("-----------------After Updating---------------------")


thisdict.pop("year")

print(thisdict)
```

```python
# The popitem() = It removes the last inserted item from the dictionary.

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}


print("-----------------Before Updating---------------------")
print(thisdict)


print("-----------------After Updating---------------------")
thisdict.popitem()

print(thisdict)
```

```python
# The del = It removes the item with the specified key name

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

print("-----------------Before Updating---------------------")
print(thisdict)

print("-----------------After Updating---------------------")
del thisdict
```

```python
# The clear() = It empties the dictionary

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

print("------------------Before Updating----------------------------")
print(thisdict)

print("------------------After Updating----------------------------")
thisdict.clear()

print(thisdict)
```

## Python - Loop Dictionaries

## Loop Through a Dictionary

You can loop through a dictionary by using a for loop.

When looping through a dictionary, the return value are the keys of the dictionary, but there are methods to return the values as well.

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}


for i in thisdict:
    print(i)
```

```python
### Looping through all the values

for i in thisdict.values():
    print(i)
```

```python
### Looping through all the items

for i in thisdict.items():
    print(i)
```

## Python - Copy Dictionaries

## Copy a Dictionary

You cannot copy a dictionary simply by typing dict2 = dict1, because: dict2 will only be a reference to dict1, and changes made in dict1 will automatically also be made in dict2.

There are ways to make a copy, one way is to use the built-in Dictionary method copy()

```python
student={"name":"Parag","age":30,"address":"India"}
print(student)
```

```python
### Original Copy

student_copy = student

print(student_copy)

print(student)
```

```python
student["age"] = 31


print(student_copy)

print(student)
```

### OBSERVATIONS:

1.  Here we are performing the original copy where any change in the original dictionary is reflected in the copied dictionary.

```python
### Shallow Copy

student1 = student.copy()

print(student)

print(student1)
```

```python
student['age'] = 30

print(student)

print(student1)
```

### OBSERVATIONS:

1. Here we are performing the shallow copy operation on the original data, where it just copies the references of the original data but does not copy the entire data.

## Python - Nested Dictionaries

## Nested Dictionaries

A dictionary can contain dictionaries, this is called nested dictionaries.

```python
# Create a dictionary that contain three dictionaries

myfamily = {
    "child1":{
        "name":"Parag"  ,
        "age" : 30      ,
        "address" : "Noida"
    },
    "child2":{
        "name" : "Subham" ,
        "age"  :  30      ,
        "address" : "Hyderabad"
    },
    "child3":{
        "name"  : "Nishant"  ,
        "age"   :  30        ,
        "address" : "Bangalore"
    }
}
```

```python
print(myfamily)
```

```python
# Create three dictionaries, then create one dictionary that will contain the other three dictionaries

employee1 = {
        "name"    :     "Parag"  ,
        "age"     :      30      ,
        "address" :     "Noida"
}


employee2 = {
        "name"    :     "Subham"  ,
        "age"     :      30       ,
        "address" :     "Hyderabad"
}

employee3 = {
        "name"    :     "Nishant"  ,
        "age"     :      30        ,
        "address" :     "Bangalore"
}



myemployee = {
    "employee1"  : employee1         ,
    "employee2"  : employee2         ,   
    "employee3"  : employee3
}

print(myemployee)
```

## Access Items in Nested Dictionaries

To access items from a nested dictionary, you use the name of the dictionaries, starting with the outer dictionary

```python
print("Age of Parag is :", myemployee["employee1"]["age"])
```

## Methods used in Dictionaries :

```python
# 1. clear() = It removes all the elements from the dictionary.

car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}


car.clear()

print(car)
```

```python
# 2. copy() = It returns a copy of the specified dictionary.


car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

thiscar = car.copy()

print(thiscar)
```

```python
# 3. fromkeys() = It returns a dictionary with the specified keys and the specified value.


# creating only keys

x = ('name','age','salary')

ans = dict.fromkeys(x)

print(ans)
```

```python
# 4. get() = It is used to obtain the value for a key.

car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}


x = car.get('model')

print(x)


y = car.get('year')

print(y)
```

```python
# 5. items() = It helps in getting all the keys and values pairs.

car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}


car.items()
```

```python
# 6. keys() = It is used to get all the keys from the dictionary.

car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}


car.keys()
```

```python
# 7. values() =It is used to get all the values from the dictionary.


car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}



car.values()
```

```python
# 8. pop() = It removes the item with the specified key.

car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

x = car.pop('model')

print(x)
```

```python
# 9. popitem() = It removes the item that was last inserted into the dictionary.

car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}


x = car.popitem()

print(x)
```

```python
# 10. setdefault() = It  returns the value of the item with the specified key.

## If the key does not exist, insert the key, with the specified value, see example below

car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}


x = car.setdefault("model","Bronco")

print(x)
```

```python
# 11. update() =It inserts the specified items to the dictionary.

car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}


car.update({"color":"white"})

print(car)
```
